**Simulasi Hukum Ohm**

Simulasi ini bertujuan untuk membantu memahami konsep dasar kelistrikan melalui Hukum Ohm serta aplikasinya pada beberapa model rangkaian sederhana. Hukum Ohm menyatakan bahwa besar arus listrik yang mengalir pada suatu penghantar berbanding lurus dengan tegangan dan berbanding terbalik dengan hambatan penghantarnya. Secara matematis dituliskan:

$$ I = \frac{V}{R} $$

dengan:
*   ùêº = arus listrik (Ampere)
*   ùëâ = beda potensial/tegangan (Volt)
*   ùëÖ = hambatan listrik (Ohm)

Hubungan antara tegangan dan arus bersifat linear, sehingga grafik
ùëâ terhadap
ùêº akan berupa garis lurus untuk hambatan yang konstan.

Selain itu, simulasi juga membahas rangkaian seri dan paralel, di mana hambatan totalnya ditentukan oleh:
$$ R_{\text{total}} = R_1 + R_2 + R_3 + \dots $$

Arus pada rangkaian seri bernilai sama pada setiap komponen, sedangkan tegangannya terbagi. Dan untuk rangkaian paralel:
$$ \frac{1}{R_{\text{total}}} = \frac{1}{R_1} + \frac{1}{R_2} + \frac{1}{R_3} + \dots $$

Tegangan pada rangkaian paralel sama pada setiap cabang, sedangkan arus terbagi.

Melalui simulasi ini, kita dapat melihat bagaimana arus berubah ketika tegangan dan hambatan diubah, bagaimana bentuk grafik hubungan V‚ÄìI untuk beberapa nilai hambatan, serta bagaimana perbandingan arus dan hambatan total pada rangkaian seri dan paralel.


```python
def hitung_arus(V, R):
    return V / R

def hitung_tegangan(I, R):
    return I * R

def hitung_hambatan(V, I):
    return V / I

print("Contoh Perhitungan Hukum Ohm")
print(f"Arus untuk V=12V, R=6Œ©: {hitung_arus(12, 6)} A")
print(f"Tegangan untuk I=2A, R=10Œ©: {hitung_tegangan(2, 10)} V")
print(f"Hambatan untuk V=20V, I=4A: {hitung_hambatan(20, 4)} Œ©")
```

```python
print("SIMULASI PERHITUNGAN HUKUM OHM")
print("1. Hitung Arus (I = V / R)")
print("2. Hitung Tegangan (V = I * R)")
print("3. Hitung Hambatan (R = V / I)")
pilihan = input("Pilih menu (1/2/3): ")

if pilihan == "1":
    print("\n--- Hitung Arus ---")
    V = float(input("Masukkan tegangan (V): "))
    R = float(input("Masukkan hambatan (Œ©): "))
    I = hitung_arus(V, R)
    print(f"Arus yang mengalir: {I} A")

elif pilihan == "2":
    print("\n--- Hitung Tegangan ---")
    I = float(input("Masukkan arus (A): "))
    R = float(input("Masukkan hambatan (Œ©): "))
    V = hitung_tegangan(I, R)
    print(f"Tegangan yang diperlukan: {V} V")

elif pilihan == "3":
    print("\n--- Hitung Hambatan ---")
    V = float(input("Masukkan tegangan (V): "))
    I = float(input("Masukkan arus (A): "))
    R = hitung_hambatan(V, I)
    print(f"Hambatan rangkaian: {R} Œ©")

else:
    print("Pilihan tidak valid!")
```

```python
import numpy as np
import matplotlib.pyplot as plt

R_values = [2, 4, 6, 8, 10]

I = np.linspace(0, 5, 100)

plt.figure(figsize=(8, 6))

for R in R_values:
    V = I * R
    plt.plot(I, V, label=f"R = {R} Œ©")

plt.title("Grafik Hubungan Tegangan (V) dan Arus (I) untuk Berbagai Hambatan")
plt.xlabel("Arus (A)")
plt.ylabel("Tegangan (V)")
plt.grid(True)
plt.legend()
plt.show()
```

*   Setiap garis menunjukkan hubungan linier ùëâ = ùêºùëÖ untuk nilai hambatan yang berbeda.
*   Semakin besar hambatan, semakin curam slope grafiknya.





```python
def hitung_rangkaian_seri(daftar_R, V_sumber):
    """
    Menghitung hambatan total, arus total, serta tegangan pada tiap hambatan
    untuk rangkaian seri.
    """
    R_total = sum(daftar_R)
    I_total = V_sumber / R_total
    tegangan_per_resistor = [I_total * R for R in daftar_R]

    return R_total, I_total, tegangan_per_resistor

print("RANGKAIAN SERI")
daftar_R = [4, 6, 10]
V_sumber = 12

R_total, I_total, tegangan = hitung_rangkaian_seri(daftar_R, V_sumber)

print(f"Hambatan total: {R_total} Œ©")
print(f"Arus total: {I_total:.2f} A")
print("Tegangan pada tiap resistor:")
for i, V in enumerate(tegangan):
    print(f"  R{i+1} = {daftar_R[i]} Œ© ‚Üí V = {V:.2f} V")
```

*   Pada rangkaian seri, arus sama pada setiap komponen.
*   Hambatan total diperoleh dari penjumlahan seluruh hambatan.
*   Tegangan terbagi sesuai nilai hambatan masing-masing resistor (hukum pembagian tegangan).


```python
def hitung_rangkaian_paralel(daftar_R, V_sumber):
    """
    Menghitung hambatan total, arus total, serta arus pada tiap cabang
    untuk rangkaian paralel.
    """
    R_total_inverse = sum(1 / R for R in daftar_R)
    R_total = 1 / R_total_inverse

    I_total = V_sumber / R_total

    arus_per_cabang = [V_sumber / R for R in daftar_R]

    return R_total, I_total, arus_per_cabang

print("RANGKAIAN PARALEL")
daftar_R = [4, 6, 12]
V_sumber = 12

R_total, I_total, arus_cabang = hitung_rangkaian_paralel(daftar_R, V_sumber)

print(f"Hambatan total: {R_total:.2f} Œ©")
print(f"Arus total: {I_total:.2f} A")
print("Arus pada tiap cabang:")
for i, I in enumerate(arus_cabang):
    print(f"  R{i+1} = {daftar_R[i]} Œ© ‚Üí I = {I:.2f} A")
```



*   Pada rangkaian paralel, tegangan pada tiap cabang sama.
*   Hambatan total didapat dari:
$$
R_{\text{total}} = \frac{1}{\sum\limits_{i=1}^{n} \frac{1}{R_i}}
$$


*   Arus terbagi ke setiap cabang sesuai besar hambatan masing-masing.







```python
import numpy as np
import matplotlib.pyplot as plt

V_sumber = 12

R1 = 4
R2 = np.linspace(2, 20, 100)

R_total_seri = R1 + R2
I_seri = V_sumber / R_total_seri

R_total_paralel = 1 / (1/R1 + 1/R2)
I_paralel = V_sumber / R_total_paralel

plt.figure(figsize=(8, 6))
plt.plot(R2, I_seri, label="Arus Total (Seri)")
plt.plot(R2, I_paralel, label="Arus Total (Paralel)")
plt.xlabel("Nilai Hambatan R‚ÇÇ (Œ©)")
plt.ylabel("Arus Total (A)")
plt.title("Perbandingan Arus Total Rangkaian Seri dan Paralel")
plt.grid(True)
plt.legend()
plt.show()
```



*   Pada rangkaian seri, bertambahnya R‚ÇÇ akan menaikkan hambatan total sehingga arus mengecil secara signifikan.
*   Pada rangkaian paralel, bertambahnya R‚ÇÇ tidak memberikan pengaruh sebesar rangkaian seri, karena resistansi total paralel lebih sensitif terhadap hambatan terkecil.
