#**Gerak Jatuh Bebas dengan Hambatan Udara**

Gerak jatuh bebas adalah suatu bentuk gerakan ketika sebuah benda dijatuhkan tanpa kecepatan awal
dan hanya dipengaruhi oleh gaya gravitasi. Pada kondisi ideal (tanpa adanya hambatan udara), percepatan
benda tetap, yaitu sebesar percepatan gravitasi:

\[
g = 9.8 \, m/s^2
\]

Namun dalam kondisi nyata, udara memberikan gaya hambat yang bergantung pada kecepatan benda.



## **Model Fisika yang Digunakan**

Pada simulasi ini, gaya yang bekerja pada benda adalah:

### 1. Gaya Gravitasi
$$
F_g = m \cdot g
$$

### 2. Gaya Hambat Udara (Drag)
$$
F_d = -k \cdot v
$$

dengan:

- \( k \) = koefisien hambatan udara  
- \( v \) = kecepatan benda  


## **Persamaan Gerak**

Berdasarkan Hukum Newton II:

$$
m \frac{dv}{dt} = -mg - kv
$$

Sehingga percepatannya adalah:

$$
a = -g - \frac{k}{m}v
$$

Jika nilai \( k = 0 \), maka:

$$
a = -g
$$

yang berarti gerakan jatuh bebas ideal tanpa hambatan udara.


```python
import math
import numpy as np
import matplotlib.pyplot as plt

def free_fall(v0, k, m, y0, dt):
    g = 9.8
    ys, vs, accs, ts = [y0], [v0], [], [0]

    while ys[-1] > 0:
        acc = -g - (k/m) * vs[-1]
        accs.append(acc)

        v_new = vs[-1] + acc * dt
        y_new = ys[-1] + v_new * dt

        vs.append(v_new)
        ys.append(y_new)
        ts.append(ts[-1] + dt)

    return ys, vs, accs, ts

```

```python
v0 = 0
m = 0.1
dt = 0.01
k_values = [0.0, 0.05, 0.1, 0.2]
colors = ['black', 'blue', 'orange', 'red']

results = {}

for k in k_values:
    results[k] = free_fall(v0, k, m, y0=100, dt=dt)
```

```python
plt.figure(figsize=(10,4))
for k, c in zip(k_values, colors):
    ys, vs, accs, ts = results[k]
    plt.plot(ts, ys, color=c, label=f"k = {k}")

plt.xlabel("Waktu (s)")
plt.ylabel("Ketinggian (m)")
plt.title("Simulasi Gerak Jatuh Bebas (Posisi vs Waktu)")
plt.legend()
plt.show()
```

## **Interpretasi Hasil Grafik**

Grafik yang dihasilkan diatas menunjukkan hubungan antara posisi terhadap waktu untuk berbagai nilai koefisien hambatan udara \( k \). Semakin besar nilai \( k \), semakin lambat benda jatuh karena gaya hambatan udara meningkat.

**Analisis Perilaku Grafik**

Dari hasil simulasi dapat diamati bahwa:

- Untuk \( k = 0 \) (tanpa hambatan udara), benda jatuh paling cepat. Grafik menunjukkan kurva parabola karena benda mengalami percepatan konstan akibat gravitasi.
  
- Semakin besar nilai \( k \), kurva posisi terhadap waktu menjadi semakin landai. Hal ini menunjukkan bahwa benda jatuh lebih lambat karena gaya hambatan udara mengurangi percepatan.

- Untuk nilai \( k \) yang besar, benda akhirnya mendekati kondisi kecepatan terminal, yaitu ketika:

$$
F_d = F_g \Rightarrow k \cdot v = m \cdot g
$$

Pada kondisi ini, percepatan menjadi nol dan benda jatuh dengan kecepatan konstan.

```python
plt.figure(figsize=(10,4))
for k, c in zip(k_values, colors):
    ys, vs, accs, ts = results[k]
    plt.plot(ts, vs, color=c, label=f"k = {k}")

plt.xlabel("Waktu (s)")
plt.ylabel("Kecepatan (m/s)")
plt.title("Kecepatan Benda Saat Jatuh")
plt.legend()
plt.show()
```

**Analisis Grafik Kecepatan vs Waktu**

Grafik diatas menunjukkan perubahan kecepatan benda selama jatuh bebas untuk berbagai nilai koefisien hambatan udara \( k \).

## Interpretasi Hasil

- Untuk \( k = 0 \)
  Kecepatan meningkat secara linear terhadap waktu karena satu-satunya gaya yang bekerja adalah gravitasi. Percepatannya konstan sebesar \( g = 9.8\, m/s^2 \).

- Untuk nilai \( k > 0 \)
  Kurva menjadi semakin landai karena gaya hambatan udara bekerja berlawanan dengan arah gerak sehingga percepatan berkurang.

- Untuk nilai \( k \) besar
  Grafik kecepatan mencapai nilai mendekati konstan. Kondisi ini dikenal sebagai:

$$
\text{Kecepatan Terminal: } v_t = \frac{m \cdot g}{k}
$$

```python
plt.figure(figsize=(10,4))
for k, c in zip(k_values, colors):
    ys, vs, accs, ts = results[k]
    plt.plot(ts[:-1], accs, color=c, label=f"k = {k}")

plt.axhline(-9.8, linestyle='--', color='gray', label="no drag")
plt.xlabel("Waktu (s)")
plt.ylabel("Percepatan (m/s)")
plt.title("Percepatan Selama Jatuh")
plt.legend()
plt.show()
```

##**Analisis Grafik Percepatan terhadap Waktu**

Grafik ini menampilkan bagaimana percepatan benda berubah terhadap waktu ketika benda dijatuhkan dengan berbagai nilai koefisien hambatan udara \( k \).

## Interpretasi Grafik

- Untuk \( k = 0 \)
  Percepatan bernilai konstan sebesar \( -9.8 \, m/s^2 \) (garis horizontal putus-putus). Ini menunjukkan kondisi ideal tanpa pengaruh hambatan udara.

- Untuk \( k > 0 \)  
  Percepatan tidak lagi konstan. Nilai percepatan awal mendekati \( -g \), namun semakin menurun seiring waktu. Hal ini terjadi karena gaya hambat udara semakin besar ketika kecepatan meningkat.

- Semakin besar nilai \( k \)
  Percepatan mendekati nilai nol, menandakan benda mulai mencapai: Kecepatan Terminal

Pada kondisi kecepatan terminal:

$$
F_g = F_d \Rightarrow mg = kv
$$

Sehingga percepatan menjadi nol, dan benda jatuh dengan kecepatan konstan.
