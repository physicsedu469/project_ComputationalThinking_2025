**Simulasi Gelombang Sinusoindal Tiga Fasa**

Gelombang sinusoidal adalah jenis gelombang yang memiliki bentuk kurva halus dan berulang secara teratur, mengikuti fungsi matematika sinus. Gelombang ini ditandai oleh pola naik dan turun yang simetris, dengan puncak (nilai maksimum) dan lembah (nilai minimum) yang muncul pada interval waktu yang tetap. Secara matematis, gelombang sinusoidal dapat dituliskan sebagai:

$$
y(t) = A \sin\ (ꞷ t + \phi)
$$


dengan:

A = amplitudo,

ꞷ = frekuensi sudut (rad/s),

t = waktu (s),

φ = fase awal (rad).

Dalam sistem gelombang sinusoidal terdapat tiga fasa yaitu IR, IS, dan IT yang berosilasi dengan bentuk gelombang, frekuensi, dan amplitudo sama, namun memiliki perbedaan fase antar fasanya. Ketiga gelombang ini memiliki perbedaan fase sebesar 120° dimana Fasa R memiliki sudut fase 0°, fasa S memiliki sudut fase 120°, dan fasa T memiliki sudut fase 240°.

Untuk fasa $I_R$ Secara matematis, dapat dituliskan sebagai:
$$
I_R(t) = A \sin\ (ꞷ t + 0^\circ)
$$

Untuk fasa $I_S$ Secara matematis, dapat dituliskan sebagai:
$$
I_S(t) = A \sin\ (ꞷ t + 120^\circ)
$$

Untuk fasa $I_T$ Secara matematis, dapat dituliskan sebagai:
$$
I_T(t) = A \sin\ (ꞷ t + 240^\circ)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
f = 50
T = 0.04
dt = 0.0005
omega = 2 * np.pi * f

t = np.arange(0, T, dt)


IR = A * np.sin(omega * t + np.deg2rad(0))
IS = A * np.sin(omega * t + np.deg2rad(120))
IT = A * np.sin(omega * t + np.deg2rad(240))

print("|t(s)| |ꞷ(rad/s)| |  IR  | |  IS | |  IT  |")
print("-"*110)

for i in range(20):
    print(f"{t[i]:.3f}\t{omega:.3f}\t\t{IR[i]:.3f}\t{IS[i]:.3f}\t{IT[i]:.3f}")

plt.figure(figsize=(12,6))
plt.plot(t, IR, label="sin IR (0°)", linewidth=2)
plt.plot(t, IS, label="sin IS (120°)", linewidth=2)
plt.plot(t, IT, label="sin IT (240°)", linewidth=2)
plt.title("GELOMBANG SINUSOIDAL TIGA FASA")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.legend()
plt.show()
```

Dalam grafik gelombang sinusoidal tiga fasa, tiga sinyal sinusoidal menunjukkan fasa IR, IS, dan IT. Ketiga gelombang ini berosilasi secara berulang seiring waktu, dan meskipun memiliki bentuk dasar yang sama, posisi puncak dan lembahnya tidak terjadi secara bersamaan karena setiap fasa memiliki pergeseran sudut tertentu. Pada data awal , IR berada pada titik nol, IS berada pada nilai positif dan IT berada pada nilai negatif, menunjukkan bahwa ketiganya berada pada posisi fase yang berbeda selama periode osilasi yang sama. Grafik juga menunjukkan pergeseran fase ini yang terus-menerus, dengan setiap gelombang mengikuti dan kemudian bergantian mencapai puncaknya. Oleh karena itu, pada grafik terlihat pola tiga kurva sinus yang bergerak secara harmonis meskipun saling bergeser. Pola ini mencerminkan ciri khas sistem tiga fasa yang berulang secara teratur.
