Gerak vertikal pada kenyataannya tidak hanya dipengaruhi oleh gaya gravitasi, tetapi juga oleh gaya hambatan udara yang bekerja berlawanan dengan arah gerak benda. Pada model hambatan udara linear, gaya hambatan diberikan oleh
$$F_d = -cv$$
sehingga percepatan benda tidak lagi konstan seperti pada kasus jatuh bebas ideal. Akibatnya, perubahan kecepatan dan posisi tidak dapat dihitung dengan rumus langsung, tetapi perlu dihitung secara bertahap menggunakan metode numerik. Simulasi ini bertujuan untuk memodelkan bagaimana kecepatan dan posisi benda berubah dari waktu ke waktu ketika dipengaruhi gravitasi dan hambatan udara linear, baik saat benda bergerak naik maupun saat jatuh menuju tanah.

Model gerak pada simulasi ini didasarkan pada dua persamaan diferensial:
$$\frac{dv}{dt} = -g - \frac{c}{m}v$$

$$\frac{dy}{dt} = v$$

Persamaan pertama menunjukkan perubahan kecepatan akibat kombinasi gravitasi dan hambatan udara. Persamaan kedua merupakan definisi dasar kinematika yang menyatakan bahwa kecepatan adalah laju perubahan posisi.


```python
import numpy as np
import matplotlib.pyplot as plt


# Parameter Fisika
m = 0.1        # massa (kg)
c = 0.05       # koefisien hambatan linear (kg/s)
g = 9.81       # gravitasi (m/s^2)

# Kondisi Awal
y0 = 0         # posisi awal (m)
v0 = 20        # kecepatan awal ke atas (m/s)

# Parameter Numerik
dt = 0.001      # langkah waktu
t_max = 5       # durasi maksimum simulasi

# Inisialisasi array
t = [0]
y = [y0]
v = [v0]

# Fungsi percepatan
def a(v):
    return -g - (c/m)*v

# Simulasi Euler-Cromer
while t[-1] < t_max:
    # update kecepatan terlebih dahulu
    v_new = v[-1] + dt * a(v[-1])
    y_new = y[-1] + dt * v_new

    # hentikan simulasi jika menyentuh tanah
    if y_new < 0:
        break

    v.append(v_new)
    y.append(y_new)
    t.append(t[-1] + dt)

# Hasil Output Numerik
print("=== HASIL SIMULASI GERAK VERTIKAL DENGAN HAMBATAN LINIER ===")
print(f"Waktu mencapai tanah : {t[-1]:.4f} s")
print(f"Ketinggian maksimum  : {max(y):.4f} m")
print(f"Kecepatan akhir      : {v[-1]:.4f} m/s")

v_terminal = -(m*g)/c
print(f"Kecepatan terminal teoritis : {v_terminal:.4f} m/s")

# Plot Grafik
plt.figure(figsize=(10,4))

# Grafik posisi
plt.subplot(1,2,1)
plt.plot(t, y)
plt.title("Posisi vs Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Posisi y (m)")
plt.grid()

# Grafik kecepatan
plt.subplot(1,2,2)
plt.plot(t, v)
plt.title("Kecepatan vs Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Kecepatan v (m/s)")
plt.grid()

plt.tight_layout()
plt.show()
```

**Grafik Posisi vs Waktu:** menunjukkan lintasan vertikal benda dari saat dilempar ke atas hingga jatuh kembali ke tanah. Kurva yang terbentuk berbentuk melengkung namun tidak simetris. Benda menghabiskan waktu yang lebih singkat untuk bergerak dari titik awal ke puncak (fase naik) dibandingkan waktu yang dibutuhkan untuk bergerak dari puncak kembali ke titik awal (fase turun). Ketidaksimetrisan ini adalah hasil dari gaya hambat udara yang selalu melawan arah gerak benda, sehingga perlambatan saat naik lebih besar daripada percepatan saat turun.

**Grafik Kecepatan vs Waktu:**  menunjukkan perubahan kecepatan benda dari kecepatan awal positif (arah ke atas) hingga kecepatan negatif (arah ke bawah). Kecepatan berkurang secara terus menerus, menembus nol saat benda berada di ketinggian maksimum. Pada fase jatuh, kecepatan benda meningkat secara negatif, tetapi peningkatannya dibatasi oleh gaya hambat. Kurva ini menunjukkan bahwa kecepatan benda saat menyentuh tanah lebih kecil (magnitude negatifnya lebih rendah) daripada kecepatan terminal teoritis, yang merupakan batas kecepatan yang akan dicapai benda jika jatuh bebas dalam waktu yang sangat lama.
