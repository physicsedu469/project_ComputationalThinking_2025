## **Simulasi Osilasi Pegas**

Menggunakan Metode Numerik Euler (Euler Method)

**1. Pendahuluan**

Gerak osilasi merupakan salah satu fenomena penting dalam fisika yang banyak ditemukan dalam kehidupan sehari-hari, seperti getaran pada pegas, ayunan bandul, atau suspensi kendaraan. Salah satu bentuk osilasi yang paling sederhana adalah osilasi harmonik sederhana (OHS), yaitu gerak berulang secara periodik yang terjadi karena adanya gaya pemulih.

Contoh nyata dari osilasi harmonik sederhana adalah pegas yang ditarik kemudian dilepaskan. Ketika pegas diregangkan atau ditekan, timbul gaya pemulih yang berusaha mengembalikan pegas ke posisi setimbang. Gaya tersebut mengikuti Hukum Hooke, yaitu gaya berbanding lurus dengan simpangan, tetapi arahnya berlawanan.

Gerakan ini dapat disimulasikan secara numerik menggunakan metode komputasi untuk memahami pola perpindahan terhadap waktu. Pada proyek ini digunakan Metode Numerik Euler (Finite Difference) untuk menghitung perubahan kecil pada posisi dan kecepatan setiap langkah waktu, kemudian divisualisasikan menggunakan grafik osilasi.

Gaya pegas mengikuti Hukum Hooke:

$$ F = -kx $$

Karena Hukum Newton II:

$$ F = ma $$

Maka:

$$ ma = -kx $$

Percepatan adalah turunan kecepatan terhadap waktu:

$$ a = \frac{dv}{dt} $$

Sehingga persamaan geraknya menjadi:

$$ \frac{dv}{dt} = -\frac{k}{m}x $$

Gerak ini menghasilkan **osilasi harmonik sederhana**, yaitu gerak periodik yang berulang-ulang.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter fisika
k = 10      # konstanta pegas (N/m)
m = 1       # massa benda (kg)
x = 5.0     # posisi awal (m)
v = 0.0     # kecepatan awal (m/s)
dt = 0.01   # langkah waktu
t_max = 10  # durasi simulasi

# List untuk menyimpan data
time = [0]
x_list = [x]

# Simulasi numerik dengan metode Finite Difference (Euler)
t = 0
while t < t_max:
    a = -(k/m) * x        # percepatan berdasarkan hukum Hooke
    v = v + a * dt        # update kecepatan
    x = x + v * dt        # update posisi
    t += dt

    time.append(t)
    x_list.append(x)

# Plot grafik
plt.figure()
plt.plot(time, x_list)
plt.title("Simulasi Osilasi Pegas (Metode Finite Difference)")
plt.xlabel("Waktu (s)")
plt.ylabel("Perpindahan (m)")
plt.grid(True)
plt.show()
```

**Hasil dan Pembahasan**

Dari hasil simulasi yang dihasilkan program, muncul sebuah grafik hubungan antara perpindahan posisi benda terhadap waktu. Grafik tersebut menunjukkan bentuk gelombang naik-turun secara periodik, yang merupakan ciri khas gerak osilasi harmonik sederhana.

Pada awal waktu, posisi benda berada pada x = 5 meter, artinya benda ditarik menjauhi posisi setimbang (x = 0). Setelah dilepaskan, benda mulai bergerak ke arah tengah karena adanya gaya pemulih pegas. Benda kemudian melewati posisi setimbang, terus bergerak ke arah berlawanan, lalu kembali lagi, sehingga terbentuk pola gerak bolak-balik. Pola ini terlihat pada grafik sebagai kurva berulang yang menyerupai gelombang sinus.

Simulasi ini memperlihatkan bagaimana pegas bekerja berdasarkan Hukum Hooke, yaitu gaya pegas selalu bekerja menuju posisi awal dan berbanding lurus dengan jarak regangan. Ketika benda berada jauh dari titik seimbang, gaya pemulih menjadi besar sehingga benda dipercepat menuju tengah. Namun ketika melewati titik tengah, gaya berubah arah dan benda kembali bergerak ke arah berlawanan.

Perpindahan posisi yang ditampilkan pada grafik menunjukkan bahwa pegas menghasilkan gerakan periodik, yaitu gerak yang berulang secara teratur dalam interval waktu tertentu. Pada simulasi ini, metode numerik Euler digunakan untuk menghitung posisi dan kecepatan setiap perubahan waktu kecil (
