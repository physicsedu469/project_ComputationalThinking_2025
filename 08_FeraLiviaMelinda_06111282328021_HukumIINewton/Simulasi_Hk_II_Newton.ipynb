```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 2       # massa (kg)
F = 10      # gaya (N)
a = F/m     # percepatan (m/s^2)
```

```python
dt = 0.01
t_max = 5
t = np.arange(0, t_max, dt)
```

```python
x = np.zeros_like(t)
v = np.zeros_like(t)
```

```python
x[0] = 0
v[0] = 0
```

```python
for i in range(len(t)-1):
    v[i+1] = v[i] + a*dt
    x[i+1] = x[i] + v[i]*dt
```

```python
plt.plot(t, x)
plt.title('Posisi vs Waktu (Hukum Newton II)')
plt.xlabel('Waktu (s)')
plt.ylabel('Posisi (m)')
plt.grid(True)
plt.show()
```

```python
plt.plot(t, v)
plt.title('Kecepatan vs Waktu (Hukum Newton II)')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan (m/s)')
plt.grid(True)
plt.show()
```
