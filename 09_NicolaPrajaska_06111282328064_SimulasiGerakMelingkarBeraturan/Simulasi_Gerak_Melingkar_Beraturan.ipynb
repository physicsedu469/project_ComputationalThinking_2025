```python
# ==============================
#  SIMULASI GERAK MELINGKAR BERATURAN
#  Mirip format simulasi gerak peluru Anda
# ==============================

import numpy as np
import matplotlib.pyplot as plt

# Parameter dasar
r = 5.0               # jari-jari (meter)
omega = 2.0           # kecepatan sudut (rad/s)
dt = 0.01             # langkah waktu
t_akhir = 10          # durasi simulasi

# Array waktu
t = np.arange(0, t_akhir, dt)

# Persiapan array posisi, kecepatan, percepatan
x, y = [], []
vx, vy = [], []
ax, ay = [], []



```

```python
# Simulasi numerik GMB
for ti in t:
    x_i = r * np.cos(omega * ti)
    y_i = r * np.sin(omega * ti)

    vx_i = -r * omega * np.sin(omega * ti)
    vy_i =  r * omega * np.cos(omega * ti)

    ax_i = -r * (omega**2) * np.cos(omega * ti)
    ay_i = -r * (omega**2) * np.sin(omega * ti)

    x.append(x_i)
    y.append(y_i)
    vx.append(vx_i)
    vy.append(vy_i)
    ax.append(ax_i)
    ay.append(ay_i)
```

```python
# ==============================
# Plot lintasan
# ==============================
plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.title("Lintasan Gerak Melingkar Beraturan")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid()
plt.axis("equal")
plt.show()

# ==============================
# Plot kecepatan dan percepatan (komponen)
# ==============================
plt.figure(figsize=(10,6))
plt.plot(t, vx, label="vx")
plt.plot(t, vy, label="vy")
plt.title("Komponen Kecepatan terhadap Waktu")
plt.xlabel("waktu (s)")
plt.ylabel("kecepatan (m/s)")
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(t, ax, label="ax")
plt.plot(t, ay, label="ay")
plt.title("Komponen Percepatan terhadap Waktu")
plt.xlabel("waktu (s)")
plt.ylabel("percepatan (m/sÂ²)")
plt.grid()
plt.legend()
plt.show()
```

```python
# ==============================
# Variasi parameter: pengaruh perubahan jari-jari
# ==============================

def simulasi_variatif(r):
    x_r = r * np.cos(omega * t)
    y_r = r * np.sin(omega * t)
    return x_r, y_r

r_values = [2, 5, 8]
plt.figure(figsize=(6,6))
for rv in r_values:
    xr, yr = simulasi_variatif(rv)
    plt.plot(xr, yr, label=f"r = {rv} m")

plt.title("Pengaruh Variasi Jari-jari terhadap Lintasan")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.axis("equal")
plt.grid()
plt.legend()
plt.show()
```

# Kesimpulan Fisika Gerak Melingkar Beraturan

 Dalam Gerak Melingkar Beraturan (GMB), sebuah benda bergerak pada lintasan berbentuk lingkaran dengan besar kecepatan tetap, tetapi arah kecepatannya selalu berubah. Perubahan arah ini membuat benda mengalami percepatan sentripetal, yaitu percepatan yang selalu mengarah menuju pusat lingkaran.

## 1. Kecepatan Linear dan Kecepatan Sudut

Kecepatan linear ð£ bergantung pada kecepatan sudut ð dan jari-jari ð:

### $$v = \omega r$$


Walaupun besar ð£ konstan, arah vektor kecepatan selalu berubah mengikuti lintasan lingkaran.

## 2. Percepatan Sentripetal

Percepatan ini tidak mengubah besar kecepatan, tetapi mengubah arah kecepatan.

Persamaan percepatan sentripetal:

### $$ a_s = \frac{v^2}{r} $$ atau $$ a_s = \omega^2 r $$

Arah $ a_s $ selalu tegak lurus dengan kecepatan dan menuju ke pusat lingkaran.

## 3. Periode dan Frekuensi Gerak

Waktu satu putaran penuh disebut periode
### $$
T = \frac{2\pi}{\omega}
$$
Banyaknya putaran per detik adalah frekuensi
### $$
f = \frac{1}{T}
$$


## 4. Pengaruh Parameter Fisika

Jika kecepatan sudut meningkat:
- Kecepatan linear **berbanding lurus** dengan kecepatan sudut:
### $$
v = \omega r
$$

- Percepatan sentripetal **berbanding lurus dengan kuadrat kecepatan sudut**:
### $$
a_s = \omega^2 r
$$


Jika jari-jari lingkaran membesar:
- Kecepatan linear **berbanding lurus** dengan jari-jari:
### $$
v = \omega r
$$

- Percepatan sentripetal juga **berbanding lurus** dengan jari-jari:
### $$
a_s = \omega^2 r
$$

**Gerak melingkar beraturan terjadi ketika sebuah benda bergerak dengan besar kecepatan tetap pada lintasan melingkar, namun tetap mengalami percepatan karena arah kecepatannya berubah secara terus-menerus. Percepatan ini selalu menuju pusat lingkaran dan besarnya ditentukan oleh jari-jari dan kecepatan sudut.**
