```python
# ==============================
#  SIMULASI GERAK MELINGKAR BERATURAN
# ==============================

import numpy as np
import matplotlib.pyplot as plt

# Parameter dasar
r = 5.0               # jari-jari (meter)
omega = 2.0           # kecepatan sudut (rad/s)
dt = 0.01             # langkah waktu
t_akhir = 10          # durasi simulasi

# List waktu
t = np.arange(0, t_akhir, dt)

# Persiapan List posisi, kecepatan, percepatan
x, y = [], []
vx, vy = [], []
ax, ay = [], []



```

```python
# Simulasi numerik GMB
for ti in t:
    x_i = r * np.cos(omega * ti)
    y_i = r * np.sin(omega * ti)

    vx_i = -r * omega * np.sin(omega * ti)
    vy_i =  r * omega * np.cos(omega * ti)

    ax_i = -r * (omega**2) * np.cos(omega * ti)
    ay_i = -r * (omega**2) * np.sin(omega * ti)

    x.append(x_i)
    y.append(y_i)
    vx.append(vx_i)
    vy.append(vy_i)
    ax.append(ax_i)
    ay.append(ay_i)
```

```python
# ==============================
# Plot lintasan
# ==============================
plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.title("Lintasan Gerak Melingkar Beraturan")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid()
plt.axis("equal")
plt.show()

# ==============================
# Plot kecepatan dan percepatan (komponen)
# ==============================
plt.figure(figsize=(10,6))
plt.plot(t, vx, label="vx")
plt.plot(t, vy, label="vy")
plt.title("Komponen Kecepatan terhadap Waktu")
plt.xlabel("waktu (s)")
plt.ylabel("kecepatan (m/s)")
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(t, ax, label="ax")
plt.plot(t, ay, label="ay")
plt.title("Komponen Percepatan terhadap Waktu")
plt.xlabel("waktu (s)")
plt.ylabel("percepatan (m/s)")
plt.grid()
plt.legend()
plt.show()
```

```python
# ==============================
# Variasi parameter: pengaruh perubahan jari-jari
# ==============================

def simulasi_variatif(r):
    x_r = r * np.cos(omega * t)
    y_r = r * np.sin(omega * t)
    return x_r, y_r

r_values = [2, 5, 8]
plt.figure(figsize=(6,6))
for rv in r_values:
    xr, yr = simulasi_variatif(rv)
    plt.plot(xr, yr, label=f"r = {rv} m")

plt.title("Pengaruh Variasi Jari-jari terhadap Lintasan")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.axis("equal")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("=== SIMULASI GERAK MELINGKAR BERATURAN ===")
print("Pilih Mode Uji:")
print("1. Uji Jari-jari (r)")
print("2. Uji Kecepatan Sudut (omega)")
print("3. Uji Kecepatan Linear (v)")
print("4. Uji Frekuensi (f)")
mode = input("Masukkan nomor mode: ")

# ===================================================================
# MODE 1 : UJI JARI-JARI r
# ===================================================================
if mode == "1":
    print("\n=== MODE 1: Uji Jari-jari (r) ===")

    r_min = float(input("Masukkan jari-jari minimum (meter): "))
    r_max = float(input("Masukkan jari-jari maksimum (meter): "))
    omega = float(input("Masukkan kecepatan sudut  (rad/s): "))

    r_values = np.linspace(r_min, r_max, 50)
    v_values = omega * r_values
    a_values = (omega**2) * r_values

    plt.plot(r_values, v_values)
    plt.xlabel("Jari-jari r (m)")
    plt.ylabel("Kecepatan Linear v (m/s)")
    plt.title("Hubungan v =  r")
    plt.grid(True)
    plt.show()

    print("\nContoh Hasil (r, v, a_sentripetal):")
    for i in range(0, 50, 10):
        print(f"r = {r_values[i]:.2f} m | v = {v_values[i]:.2f} m/s | a = {a_values[i]:.2f} m/s")


# ===================================================================
# MODE 2 : UJI KECEPATAN SUDUT 
# ===================================================================
elif mode == "2":
    print("\n=== MODE 2: Uji Kecepatan Sudut (omega) ===")

    omega_min = float(input("Masukkan  minimum (rad/s): "))
    omega_max = float(input("Masukkan  maksimum (rad/s): "))
    r = float(input("Masukkan jari-jari r (meter): "))

    omega_values = np.linspace(omega_min, omega_max, 50)
    v_values = omega_values * r
    a_values = (omega_values ** 2) * r

    plt.plot(omega_values, v_values)
    plt.xlabel("Kecepatan Sudut  (rad/s)")
    plt.ylabel("Kecepatan Linear v (m/s)")
    plt.title("Hubungan v =  r")
    plt.grid(True)
    plt.show()

    print("\nContoh Hasil (, v, a_sentripetal):")
    for i in range(0, 50, 10):
        print(f" = {omega_values[i]:.2f} rad/s | v = {v_values[i]:.2f} m/s | a = {a_values[i]:.2f} m/s")


# ===================================================================
# MODE 3 : UJI KECEPATAN LINEAR v
# ===================================================================
elif mode == "3":
    print("\n=== MODE 3: Uji Kecepatan Linear (v) ===")

    v_min = float(input("Masukkan v minimum (m/s): "))
    v_max = float(input("Masukkan v maksimum (m/s): "))
    r = float(input("Masukkan jari-jari r (meter): "))

    v_values = np.linspace(v_min, v_max, 50)
    omega_values = v_values / r
    a_values = (v_values**2) / r

    plt.plot(v_values, a_values)
    plt.xlabel("Kecepatan Linear v (m/s)")
    plt.ylabel("Percepatan Sentripetal a (m/s)")
    plt.title("Hubungan a = v / r")
    plt.grid(True)
    plt.show()

    print("\nContoh Hasil (v, , a_sentripetal):")
    for i in range(0, 50, 10):
        print("v = {v_values[i]:.2f} m/s |  = {omega_values[i]:.2f} rad/s | a = {a_values[i]:.2f} m/s")


# ===================================================================
# MODE 4 : UJI FREKUENSI f
# ===================================================================
elif mode == "4":
    print("\n=== MODE 4: Uji Frekuensi (f) ===")

    f_min = float(input("Masukkan frekuensi minimum (Hz): "))
    f_max = float(input("Masukkan frekuensi maksimum (Hz): "))
    r = float(input("Masukkan jari-jari r (meter): "))

    f_values = np.linspace(f_min, f_max, 50)
    T_values = 1 / f_values
    omega_values = 2 * np.pi * f_values
    v_values = 2 * np.pi * r * f_values

    plt.plot(f_values, v_values)
    plt.xlabel("Frekuensi f (Hz)")
    plt.ylabel("Kecepatan Linear v (m/s)")
    plt.title("Hubungan v = 2 r f")
    plt.grid(True)
    plt.show()

    print("\nContoh Hasil (f, T, , v):")
    for i in range(0, 50, 10):
        print(f"f = {f_values[i]:.2f}  Hz | T = {T_values[i]:.3f} s |  = {omega_values[i]:.2f} rad/s | v = {v_values[i]:.2f} m/s")


# ===================================================================
else:
    print("Mode tidak valid! Pilih 1, 2, 3 atau 4.")
```
