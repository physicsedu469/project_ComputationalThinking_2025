# **Simulasi Hukum Bernoulli**

Hukum Bernoulli menyatakan bahwa kenaikan kecepatan aliran dari fluida mampu menyebabkan adanya penurunan tekanan fluida secara bersamaan. Atau juga bisa diartikan dengan menurunnya energi potensial fluida tersebut. Persamaan Bernoulli dapat dituliskan sebagai:

$$P+\frac{1}{2}\rho v^2 + \rho g h$$

**Efek Venturi**

Efek Venturi dalam aliran fluida adalah fenomena di mana aliran fluida dipercepat saat melewati penyempitan, sering digunakan dalam aplikasi termal dan mekanik.

Fluida mengalir melalui pipa dengan dua penampang, A1 dan A2, di mana A1 > A2. Menggunakan Prinsip Kontinuitas, kita tahu bahwa:
$$A_1 v_1 = A_2 v_2$$

Dan dengan prinsip Bernoulli, perbedaan tekanan dapat dihitung sebagai:

$$P_1 - P_2 = \frac{1}{2}\rho (v_2^2 - v_1^2)$$

Dengan demikian, kita dapat menghitung kecepatan dan tekanan pada berbagai titik dalam pipa sesuai dengan perubahan luas penampang.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter awal
A1 = 0.01
v1 = 2

A2 = np.linspace(0.002, 0.01, 200)

v2 = (A1 * v1) / A2

# Plot grafik kecepatan vs luas penampang
plt.figure(figsize=(8, 5))
plt.plot(A2, v2, linewidth=2)

plt.title("Grafik Kecepatan vs Luas Penampang (Efek Venturi)")
plt.xlabel("Luas Penampang (m²)")
plt.ylabel("Kecepatan Fluida (m/s)")
plt.grid(True)

plt.gca().invert_xaxis()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter fluida dan kondisi awal
rho = 1000
A1 = 0.01
v1 = 2
P1 = 101325


A2 = np.linspace(0.002, 0.01, 200)


v2 = (A1 * v1) / A2

# Menghitung tekanan:
P2 = P1 - 0.5 * rho * (v2**2 - v1**2)

# Plot grafik tekanan terhadap luas penampang
plt.figure(figsize=(8, 5))
plt.plot(A2, P2, linewidth=2)

plt.title("Grafik Tekanan vs Luas Penampang (Efek Venturi)")
plt.xlabel("Luas Penampang (m²)")
plt.ylabel("Tekanan (Pa)")
plt.grid(True)
plt.gca().invert_xaxis()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter fluida dan kondisi awal
rho = 1000
A1 = 0.01
v1 = 2
P1 = 101325


A2 = np.linspace(0.002, 0.01, 200)


v2 = (A1 * v1) / A2


P2 = P1 - 0.5 * rho * (v2**2 - v1**2)

# Plot grafik Tekanan vs Kecepatan
plt.figure(figsize=(8, 5))
plt.plot(v2, P2, linewidth=2)

plt.title("Grafik Tekanan vs Kecepatan (P–v) — Efek Venturi")
plt.xlabel("Kecepatan Fluida (m/s)")
plt.ylabel("Tekanan (Pa)")
plt.grid(True)
plt.show()
```

# Kesimpulan

Berdasarkan simulasi dan grafik yang dibuat, dapat disimpulkan bahwa Efek Venturi terjadi karena dua prinsip utama, yaitu Persamaan Kontinuitas dan Hukum Bernoulli. Saat fluida mengalir melalui pipa yang menyempit, luas penampang yang lebih kecil memaksa fluida bergerak lebih cepat agar alirannya tetap sama. Hal ini terlihat jelas pada grafik Kecepatan vs Luas Penampang, di mana kecepatan naik tajam ketika pipa makin sempit.

Ketika kecepatan fluida meningkat, tekanannya justru menurun. Grafik Tekanan vs Luas Penampang menunjukkan bahwa daerah yang sempit memiliki tekanan yang lebih rendah dibandingkan bagian yang lebih lebar. Ini sejalan dengan Hukum Bernoulli, yang menyatakan bahwa energi kinetik yang bertambah harus dibayar dengan penurunan tekanan.

Hubungan terbalik antara kecepatan dan tekanan semakin terlihat pada grafik Tekanan vs Kecepatan (P–v). Grafik tersebut menurun, yang berarti semakin cepat fluida bergerak, semakin kecil tekanannya. Pola khas ini adalah ciri utama Efek Venturi dan banyak dimanfaatkan dalam kehidupan sehari-hari, misalnya pada venturimeter, karburator, alat penyemprot parfum, nebulizer, hingga prinsip kerja sayap pesawat.
