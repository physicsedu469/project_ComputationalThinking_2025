## A.	PENDAHULUAN
Medan magnet merupakan salah satu besaran fisis yang sangat penting dalam memahami interaksi elektromagnetik di sekitar benda yang dialiri arus listrik. Ketika arus mengalir melalui sebuah kawat konduktor, medan magnet akan terbentuk mengelilingi kawat tersebut dengan arah melingkar sesuai kaidah tangan kanan. Fenomena pembentukan medan magnet oleh arus listrik ini pertama kali dijelaskan oleh Hans Christian √òrsted, dan hingga kini menjadi dasar dalam banyak aplikasi teknologi seperti motor listrik, generator, transformator, sensor arus, serta berbagai perangkat elektromagnetik lainnya. Keberadaan medan magnet ini bergantung pada besar arus listrik serta jarak pengamatan dari kawat, sehingga analisis hubungan antara medan magnet dan jaraknya sangat penting untuk memahami perilaku elektromagnetik secara kuantitatif.

Secara teoritis, besar medan magnet pada jarak tertentu dari kawat lurus panjang yang dialiri arus dapat dinyatakan melalui rumus:

$$
B(r) = \frac{\mu_0 I}{2\pi r}
$$

Rumus ini menunjukkan bahwa medan magnet B berbanding terbalik dengan jarak $r$ dari kawat. Semakin dekat suatu titik ke kawat yang berarus, maka medan magnet yang ditimbulkan semakin besar, sedangkan semakin jauh dari kawat, medan magnet semakin kecil. Konstanta $\mu_0$ yang dikenal sebagai permeabilitas vakum, bernilai $4\pi \times 10^{-7}\ \text{T}$. Persamaan ini merupakan bentuk sederhana dari hukum Biot Savart apabila kawat dianggap sangat panjang dan arus mengalir secara seragam.

Analisis medan magnet akibat kawat lurus berarus tidak hanya penting dari sisi teori, tetapi juga dalam pemodelan dan perhitungan fisika terapan. Dengan mengetahui bagaimana medan magnet berubah terhadap jarak, berbagai aplikasi teknik dapat dirancang dengan lebih akurat, terutama pada sistem kelistrikan dan komponen elektromagnetik. Selain itu, persamaan matematis yang sederhana namun fundamental ini memberikan dasar yang kuat untuk mempelajari fenomena elektromagnetik lebih lanjut, seperti induksi elektromagnetik, fluks magnet, serta interaksi antara medan magnet dan materi. Oleh karena itu, kajian terhadap medan magnet kawat lurus berarus menjadi topik penting dalam fisika dasar maupun dalam pengembangan teknologi modern.

## B.	METODE TRAPEZOIDAL
Metode trapezoidal merupakan salah satu teknik integrasi numerik yang digunakan untuk menghampiri nilai integral dari suatu fungsi yang tidak mudah diselesaikan secara analitik. Teknik ini bekerja dengan membagi interval integrasi menjadi beberapa segmen kecil dengan lebar yang sama, kemudian setiap segmen dianggap membentuk sebuah trapesium. Luas total dari seluruh trapesium tersebut kemudian dijumlahkan untuk memperoleh nilai pendekatan integral. Metode ini termasuk dalam keluarga Newton Cotes formula orde satu, yang menggunakan interpolasi linear untuk menghampiri kurva fungsi di sepanjang interval integrasi.

Secara umum, integral suatu fungsi pada interval $[a, b]$
dapat dituliskan sebagai:
$$\int_{a}^{b} f(x)\, dx$$

Namun, jika fungsi tersebut tidak dapat diintegralkan secara eksak atau sulit dievaluasi, maka diperlukan pendekatan numerik. Dalam metode trapezoidal, interval $[a, b]$
dibagi menjadi $ n $ titik pembagi sehingga diperoleh lebar segmen:
$$h = \frac{b - a}{n - 1}$$

Dengan demikian, titik-titik pembagi dapat dituliskan sebagai:

$$x_i = a + i\,h,\quad \text{untuk } i = 0,1,2,\ldots,n-1$$

Rumus utama metode trapezoidal untuk menghampiri integral adalah:
$$
\int_{a}^{b} f(x)\, dx
\approx
\frac{h}{2} \left[\, f(x_0) + 2\sum_{i=1}^{n-1} f(x_i) + f(x_n) \,\right]
$$

Rumus tersebut memberikan nilai pendekatan dengan menganggap area di bawah kurva fungsi sebagai kumpulan trapesium yang luasnya dihitung satu per satu. Semakin besar jumlah segmen $n$ semakin kecil nilai galat (error) yang dihasilkan dan semakin mendekati nilai integral sebenarnya.

Galat metode trapezoidal secara teoritis dapat dituliskan dengan rumus:
$$
E_T = -\,\frac{(b-a)^3}{12\,n^{2}}\, f''(\xi)
$$
dimana $f''(\xi)$ adalah turunan kedua dari fungsi pada suatu titik $\xi$ di interval $[ùëé,ùëè]$

Rumus ini menunjukkan bahwa galat metode trapezoidal berbanding terbalik dengan kuadrat jumlah segmen. Artinya, semakin besar jumlah titik pembagi, semakin akurat hasilnya.

Dalam konteks perhitungan medan magnet akibat kawat lurus berarus, metode trapezoidal digunakan untuk menentukan nilai integral dari fungsi:

$$
B(r) = \frac{\mu_0 I}{2\pi r}
$$

Fungsi ini memiliki bentuk monoton menurun tanpa osilasi, sehingga sangat cocok dihampiri menggunakan metode trapezoidal. Dengan membagi rentang jarak r ke dalam segmen-segmen kecil, nilai medan magnet dapat dihitung secara bertahap dan diintegrasikan dengan stabil serta akurat. Oleh karena itu, metode trapezoidal menjadi pilihan yang efektif dan efisien untuk menghitung karakteristik medan magnet secara komputasi.


## C.	ALGORITMA DENGAN MENGGUNAKAN GOOGLE COLAB
Algoritma komputasi disusun untuk menghitung dan memvisualisasikan besar medan magnet akibat kawat lurus berarus menggunakan bahasa pemrograman google colab.

Berikut adalah  kode colab yang digunakan dalam algoritma ini:

a)	Bagaimana bentuk grafik hubungan antara jarak $r$ terhadap besar medan magnet $B(r)$? Jelaskan berdasarkan grafik yang dihasilkan dari kode colab.


```python
import numpy as np
import matplotlib.pyplot as plt

# nilai konstanta mu0 (permeabilitas vakum)
mu0 = 4*np.pi*1e-7

#arus listrik yang lewat di kawat
I = 5
# fungsi untuk hitung B(r)
def B(r):
    return mu0 * I / (2*np.pi * r)

#rentang jaraknya
#dari 0.01 meter sampe 0.1 meter
#saya pakai 1000 titik biar grafiknya halus
r_vals = np.linspace(0.01, 0.1, 1000)

#hitung nilai B nya untuk setiap r tadi
B_vals = B(r_vals)

#grafik
plt.plot(r_vals, B_vals)

# label sumbu grafik
plt.xlabel("Jarak r (m)")
plt.ylabel("Medan Magnet B (T)")

# judul grafik
plt.title("Grafik Medan Magnet terhadap Jarak pada Kawat Berarus")
```

Grafik menunjukkan bahwa besar medan magnet $B(r)$ menurun tajam ketika jarak $r$ bertambah, sesuai persamaan

$$B(r) = \frac{\mu_0 I}{2\pi r}$$

Pada jarak yang sangat dekat dengan kawat, medan magnet bernilai besar karena nilai $r$ kecil, namun ketika $r$ diperbesar, medan magnet turun secara signifikan mengikuti pola hiperbolik menurun. Visualisasi ini memperlihatkan hubungan berbanding terbalik antara $B$ dan $r$ serta menguatkan konsep fisika bahwa medan magnet kawat lurus menyebar semakin lemah seiring bertambahnya jarak dari sumber arus.

b)	Bandingkan hasil metode trapezoidal dan hasil analitik menggunakan kode Python. Tunjukkan selisihnya ($error$).


```python
# fungsi trapezoidal untuk menghitung integral secara numerik.
def trapezoidal(f, a, b, n):
    #titik-titik dari batas a sampe b
    x = np.linspace(a, b, n)

    #nilai fungsi B(r) di tiap titik x
    y = f(x)

    # Hitung lebar langkahnya, h
    h = (b - a) / (n - 1)

    # Rumus trapezoidal: h * (0.5*y0 + y1 + ... + yn-1 + 0.5*yn)
    integral = h * (0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1])
    return integral


#hitung integralnya secara numerik
# f = B (fungsi medan magnet), a=0.01, b=0.1, n=1000
hasil_numerik = trapezoidal(B, 0.01, 0.1, 1000)
print("Hasil numerik =", hasil_numerik)

#rumus hasil analitiksesuai teori integral B(r)
hasil_analitik = (mu0*I/(2*np.pi)) * np.log(0.1/0.01)
print("Hasil analitik =", hasil_analitik)

#selisih antara keduanya
error = abs(hasil_analitik - hasil_numerik)
print("Error =", error)
```

Hasil perhitungan menunjukkan bahwa nilai numerik dan analitik sangat berdekatan dengan error sekitar $6.69583 \times 10^{-12}$ yang berarti perbedaannya hampir tidak signifikan. Nilai ini membuktikan bahwa metode trapezoidal sangat akurat untuk menghitung integral fungsi $\frac{1}{r}$ pada interval $0.01‚Äì0.1$ m, terutama ketika jumlah pembagian n cukup besar $(1000$ titik). Selisih yang sangat kecil ini memperlihatkan bahwa pendekatan numerik mampu meniru hasil analitik dengan presisi tinggi, sehingga valid untuk digunakan dalam perhitungan fisika komputasi.


c)	Bagaimana pengaruh jumlah pembagian $(n)$ terhadap besar $error$ metode trapezoidal? Tampilkan grafik error terhadap $n$.

```python
import numpy as np
import matplotlib.pyplot as plt  # ini buat grafiknya

#nilai mu0 (permeabilitas vakum)
mu0 = 4*np.pi*1e-7

#nilai arusnya, saya samakan dengan yang sebelumnya
I = 5

# fungsi menghitung  medan magnet B(r)
# masukin r nanti keluar nilai B-nya
def B(r):
    return mu0 * I / (2*np.pi * r)

# fungsi trapezoidal untuk menghitung integral numerik
def trapezoidal(f, a, b, n):
    x = np.linspace(a, b, n)   #titik-titiknya
    y = f(x)                   # nilai fungsi di titik tersebut
    h = (b - a) / (n - 1)      # lebar langkah tiap trapesium
    integral = h*(0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1])
    return integral

#menghitung hasil analitiknya (nilai asli)
hasil_analitik = (mu0*I/(2*np.pi)) * np.log(0.1/0.01)

# nilai n (jumlah pembagian)untuk melihat pengaruhnya terhadap error
n_vals = [10, 50, 100, 200, 500, 1000, 2000]
errors = []   # tempat nyimpen errornya

# menghitung error buat tiap n
for n in n_vals:
    hasil_numerik = trapezoidal(B, 0.01, 0.1, n)
    error = abs(hasil_analitik - hasil_numerik)  # selisihnya
    errors.append(error)

# grafik error vs n
plt.plot(n_vals, errors)
plt.xlabel("Jumlah Pembagian (n)")
plt.ylabel("Error (|analitik - numerik|)")
plt.title("Grafik Error Metode Trapezoidal untuk Berbagai n")
plt.grid(True)
plt.show()

```

Grafik $error$ terhadap $n$ menunjukkan bahwa semakin besar jumlah pembagian $(n)$, semakin kecil nilai $error$ yang dihasilkan metode trapezoidal. Pada $n$ kecil, error masih cukup besar karena bentuk fungsi $\frac{1}{r}$ yang melengkung tidak dapat diikuti dengan baik oleh segmen trapezoid. Namun, ketika $n$ diperbesar hingga ratusan atau ribuan, nilai $error$ menurun tajam karena langkah integrasi menjadi semakin kecil sehingga kurva fungsi dapat didekati lebih akurat. Hasil ini membuktikan bahwa akurasi metode trapezoidal sangat bergantung pada jumlah titik pembagian yang digunakan.
