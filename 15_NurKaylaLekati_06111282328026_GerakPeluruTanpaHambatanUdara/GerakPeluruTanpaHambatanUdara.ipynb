**A. Pendahuluan**

Gerak peluru merupakan salah satu topik fundamental dalam kajian mekanika klasik yang membahas gerak suatu benda yang dilemparkan ke udara dengan kecepatan awal tertentu dan kemudian bergerak di bawah pengaruh gaya gravitasi. Dalam kondisi ideal, gerak peluru sering dianalisis dengan mengabaikan hambatan udara, sehingga satu-satunya gaya yang bekerja pada benda adalah gaya gravitasi yang arahnya vertikal ke bawah. Model ini banyak digunakan dalam pembelajaran fisika karena mampu menggambarkan konsep dasar hubungan antara posisi, kecepatan, dan percepatan secara sederhana namun representatif.

Secara analitik, gerak peluru tanpa hambatan udara memiliki solusi matematis yang relatif mudah diturunkan. Posisi benda sebagai fungsi waktu dapat diperoleh langsung melalui persamaan kinematika dengan percepatan konstan. Namun, dalam konteks berpikir komputasi dan pemodelan numerik, penyelesaian analitik tidak selalu menjadi fokus utama. Pendekatan numerik diperlukan untuk melatih kemampuan memformulasikan permasalahan fisika ke dalam bentuk algoritma, serta untuk memahami bagaimana solusi sistem fisis dapat diperoleh secara bertahap melalui proses komputasi.

Pendekatan numerik menjadi sangat penting terutama ketika sistem yang dikaji lebih kompleks, misalnya melibatkan gaya tambahan, kondisi awal yang bervariasi, atau ketika solusi analitik sulit diperoleh. Oleh karena itu, mempelajari metode numerik pada kasus gerak peluru yang sederhana dapat menjadi jembatan konseptual sebelum diterapkan pada sistem dinamis yang lebih rumit.

Gerak peluru dapat dimodelkan menggunakan persamaan diferensial orde satu yang menyatakan hubungan antara posisi dan kecepatan, serta antara kecepatan dan percepatan. Persamaan dasar gerak peluru tanpa hambatan udara dinyatakan sebagai berikut:

$$
\frac{dx}{dt} = v_x
$$

$$

$$

**B. METODE NUMERIK GERAK PELURU TANPA HAMBATAN UDARA**

Gerak peluru tanpa hambatan udara merupakan masalah dinamika klasik yang dipengaruhi oleh gaya gravitasi konstan. Dengan memilih sumbu horizontal $x$ dan vertikal $y$, serta mengabaikan hambatan udara, gaya yang bekerja pada benda hanyalah gaya gravitasi:

$$
\vec{F} = m\vec{g}
$$

Berdasarkan hukum II Newton:
$$
\vec{F} = m\vec{a}
$$

Sehingga diperoleh percepatan:
$$
\vec{a} = \vec{g}
$$

Jika sistem koordinat didefinisikan dengan sumbu $x$ sebagai arah horizontal dan sumbu $y$ sebagai arah vertikal ke atas, maka percepatan pada masing-masing sumbu adalah:
$$
a_x = 0
$$
$$
a_y = -g
$$

Komponen kecepatan didefinisikan sebagai turunan pertama posisi terhadap waktu:
$$
v_x = \frac{dx}{dt}, \quad v_y = \frac{dy}{dt}
$$

Sedangkan percepatan merupakan turunan pertama kecepatan terhadap waktu:
$$
\frac{dv_x}{dt} = a_x = 0
$$
$$
\frac{dv_y}{dt} = a_y = -g
$$

Dengan demikian, persamaan diferensial orde dua gerak peluru dapat diubah menjadi sistem persamaan diferensial orde satu sebagai berikut:
$$
\frac{dx}{dt} = v_x
$$
$$
\frac{dy}{dt} = v_y
$$
$$
\frac{dv_x}{dt} = 0
$$
$$
\frac{dv_y}{dt} = -g
$$

Karena solusi analitik tidak selalu praktis untuk divisualisasikan secara komputasi diskrit, maka digunakan metode numerik, yaitu metode Euler dan metode Runge窶適utta orde empat (RK4).

---

**Metode Euler**

Metode Euler diperoleh dari pendekatan turunan pertama:

$$
\frac{dy}{dt} \approx \frac{y_{n+1} - y_n}{h}
$$

sehingga diperoleh skema iteratif umum:

$$
y_{n+1} = y_n + h f(y_n)
$$

Untuk gerak peluru tanpa hambatan udara, skema Euler diterapkan pada posisi dan kecepatan sebagai berikut:

$$
x_{n+1} = x_n + h v_{x,n}
$$

$$
y_{n+1} = y_n + h v_{y,n}
$$

$$
v_{x,n+1} = v_{x,n}
$$

$$
v_{y,n+1} = v_{y,n} - g h
$$

Metode Euler memiliki bentuk sederhana, tetapi akurasinya bergantung pada besar langkah waktu $h$.

---

**Metode Runge窶適utta Orde Empat (RK4)**

Untuk sistem persamaan diferensial umum:

$$
\frac{dy}{dt} = f(y,t)
$$

metode Runge窶適utta orde empat didefinisikan melalui empat tahap kemiringan:

$$
k_1 = h f(y_n)
$$

$$
k_2 = h f\left(y_n + \frac{1}{2}k_1\right)
$$

$$
k_3 = h f\left(y_n + \frac{1}{2}k_2\right)
$$

$$
k_4 = h f(y_n + k_3)
$$

Solusi pada langkah berikutnya diberikan oleh:

$$
y_{n+1} = y_n + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4)
$$

Pada gerak peluru tanpa hambatan udara, metode RK4 diterapkan secara simultan pada variabel posisi dan kecepatan $(x, y, v_x, v_y)$ untuk memperoleh lintasan gerak yang lebih akurat dibandingkan metode Euler.


**C. Algoritma dengan Menggunakan Python**

**Nilai $h$ = 0.01**


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Parameter fisika dan numerik
g = 9.8          # percepatan gravitasi (m/s^2)
v0 = 30.0        # kecepatan awal (m/s)
theta_deg = 45.0 # sudut elevasi (derajat)
h = 0.01         # langkah waktu (s)
t_max = 5.0      # waktu simulasi maksimum (s)

theta = np.deg2rad(theta_deg)  # konversi ke radian

# 2. Kondisi awal
x0 = 0.0
y0 = 0.0
vx0 = v0 * np.cos(theta)
vy0 = v0 * np.sin(theta)

# 3. Fungsi turunan untuk sistem Y = [x, y, vx, vy]
def f(Y, t):
    x, y, vx, vy = Y
    dxdt = vx
    dydt = vy
    dvxdt = 0.0
    dvydt = -g
    return np.array([dxdt, dydt, dvxdt, dvydt])

# 4. Metode Euler
def euler_projectile(x0, y0, vx0, vy0, h, t_max):
    t_values = [0.0]
    x_values = [x0]
    y_values = [y0]
    vx_values = [vx0]
    vy_values = [vy0]

    t = 0.0
    x, y, vx, vy = x0, y0, vx0, vy0

    while t < t_max and y >= 0:
        # hitung turunan
        dxdt = vx
        dydt = vy
        dvxdt = 0.0
        dvydt = -g

        # update dengan Euler
        x = x + h * dxdt
        y = y + h * dydt
        vx = vx + h * dvxdt
        vy = vy + h * dvydt
        t = t + h

        t_values.append(t)
        x_values.append(x)
        y_values.append(y)
        vx_values.append(vx)
        vy_values.append(vy)

    return np.array(t_values), np.array(x_values), np.array(y_values)

# 5. Metode Runge窶適utta Orde 4 (RK4)
def rk4_projectile(x0, y0, vx0, vy0, h, t_max):
    t_values = [0.0]
    x_values = [x0]
    y_values = [y0]

    t = 0.0
    Y = np.array([x0, y0, vx0, vy0])

    while t < t_max and Y[1] >= 0:
        k1 = h * f(Y, t)
        k2 = h * f(Y + 0.5 * k1, t + 0.5 * h)
        k3 = h * f(Y + 0.5 * k2, t + 0.5 * h)
        k4 = h * f(Y + k3, t + h)

        Y = Y + (k1 + 2*k2 + 2*k3 + k4) / 6.0
        t = t + h

        t_values.append(t)
        x_values.append(Y[0])
        y_values.append(Y[1])

    return np.array(t_values), np.array(x_values), np.array(y_values)

# 6. Menjalankan simulasi
t_euler, x_euler, y_euler = euler_projectile(x0, y0, vx0, vy0, h, t_max)
t_rk4, x_rk4, y_rk4 = rk4_projectile(x0, y0, vx0, vy0, h, t_max)

# 7. Visualisasi lintasan
plt.figure(figsize=(8, 5))
plt.plot(x_euler, y_euler, label="Euler")
plt.plot(x_rk4, y_rk4, linestyle="--", label="RK4")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Lintasan Gerak Peluru Tanpa Hambatan Udara\nMetode Euler vs RK4")
plt.legend()
plt.grid(True)
plt.show()
```

**Solusi Analitik tanpa Metode Numerik**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter fisika
v0 = 20          # kecepatan awal (m/s)
theta = 45       # sudut (derajat)
g = 9.8          # gravitasi (m/s^2)

# Konversi sudut ke radian
theta_rad = np.radians(theta)

# Waktu simulasi
t = np.linspace(0, 3, 200)

# Persamaan analitik gerak peluru
x = v0 * np.cos(theta_rad) * t
y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2

# Plot lintasan
plt.figure(figsize=(6,4))
plt.plot(x, y, label='Solusi Analitik')
plt.axhline(0)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Gerak Peluru Tanpa Hambatan Udara (Analitik)')
plt.legend()
plt.grid()
plt.show()
```
