# **Pemodelan Komputasi Gerak Osilasi Teredam dengan Pendekatan Penyelesaian Numerik**
**A. Pendahuluan**

Gerak osilasi merupakan salah satu topik mendasar dalam fisika yang banyak ditemukan pada berbagai sistem nyata, mulai dari getaran pegas, ayunan sederhana, hingga resonansi pada rangkaian listrik. Model paling dasar yang digunakan untuk menggambarkan fenomena tersebut adalah osilator harmonik, yaitu sistem yang bergerak di bawah pengaruh gaya pemulih yang sebanding dengan perpindahan dari titik setimbang. Dalam kondisi ideal tanpa adanya kehilangan energi, sistem ini menghasilkan gerak periodik yang dapat dijelaskan secara analitik menggunakan fungsi trigonometri.

Namun pada sistem nyata, gaya gesek atau resistansi medium menyebabkan energi mekanik sistem berkurang seiring waktu. Fenomena ini dikenal sebagai osilasi teredam (damped harmonic motion). Kehadiran gaya redaman membuat gerak sistem menjadi lebih kompleks dibandingkan osilator harmonik sederhana karena amplitudo tidak lagi konstan, melainkan menyusut secara bertahap. Bergantung pada besar redaman, sistem dapat berada dalam tiga regime: underdamped, critically damped, atau overdamped, masing-masing dengan karakteristik dinamika yang berbeda.

Secara matematis, gerak osilator harmonik teredam digambarkan oleh persamaan diferensial orde dua yang sulit diselesaikan secara analitik untuk kondisi tertentu, terutama ketika sistem dipengaruhi gaya eksternal atau memiliki parameter yang bervariasi terhadap waktu. Oleh karena itu, penyelesaian numerik menjadi pendekatan utama dalam menganalisis dinamika osilasi pada berbagai kasus. Metode numerik memungkinkan kita memprediksi perubahan posisi dan kecepatan secara bertahap menggunakan skema integrasi berbasis langkah waktu, seperti metode Euler maupun Runge–Kutta.

Dalam proyek ini, dilakukan pemodelan dan simulasi osilator harmonik teredam menggunakan bahasa pemrograman Python. Persamaan gerak diturunkan mulai dari gaya-gaya yang bekerja pada sistem, kemudian ditransformasikan ke bentuk sistem persamaan diferensial orde satu agar dapat diselesaikan secara numerik. Selanjutnya digunakan metode integrasi Runge–Kutta orde empat (RK4) untuk memperoleh evolusi posisi dan kecepatan sebagai fungsi waktu. Hasil simulasi divisualisasikan dalam bentuk grafik sehingga perubahan amplitudo, karakteristik osilasi, dan pengaruh redaman dapat diamati secara jelas. Studi ini memberikan pemahaman komprehensif mengenai dinamika osilasi teredam sekaligus menunjukkan peran penting komputasi dalam menganalisis sistem fisika yang tidak selalu dapat diselesaikan secara analitik.

# **B. METODE NUMERIK GERAK OSILASI**

Pemodelan gerak osilasi teredam berawal dari analisis gaya-gaya yang bekerja pada sebuah benda bermassa (m) yang terhubung dengan pegas. Berbeda dengan osilasi harmonik ideal, sistem fisik nyata selalu mengalami kehilangan energi akibat adanya interaksi dengan medium di sekitarnya. Kehilangan energi ini muncul dalam bentuk gaya redaman yang bekerja melawan arah gerak benda. Kombinasi antara gaya pemulih dan gaya redaman inilah yang menentukan karakteristik dari gerak osilasi teredam.

Gerak benda dipengaruhi oleh dua gaya utama. Pertama, gaya pemulih yang diberikan oleh pegas mengikuti Hukum Hooke, yang menyatakan bahwa gaya tersebut sebanding dengan perpindahan benda dari titik setimbang dan selalu mengarah kembali ke titik tersebut. Secara matematis, gaya pemulih dirumuskan sebagai:

$$F_{\text{pegas}} = -kx$$

Selain itu, benda juga mengalami gaya redaman yang ditimbulkan oleh gesekan fluida atau resistansi medium. Besarnya gaya redaman sebanding dengan kecepatan gerak:

$$F_{\text{redaman}} = -b\frac{dx}{dt}$$

Menurut Hukum II Newton, jumlah semua gaya yang bekerja pada benda sama dengan massa dikalikan percepatannya. Dengan menjumlahkan gaya pemulih dan gaya redaman, diperoleh persamaan gerak sistem:

$$m\frac{d^2x}{dt^2} = -kx - b\frac{dx}{dt}$$

yang dapat ditulis dalam bentuk umum:

$$m\frac{d^2x}{dt^2} + b\frac{dx}{dt} + kx = 0\tag{1}$$

Persamaan (1) merupakan persamaan diferensial orde dua yang menggambarkan dinamika osilasi teredam. Amplitudo gerak tidak lagi konstan, melainkan berkurang seiring waktu akibat gaya redaman. Namun, integrator numerik seperti Euler atau Runge–Kutta tidak dapat langsung bekerja pada persamaan orde dua sehingga model perlu ditransformasikan ke bentuk sistem persamaan diferensial orde satu.

Transformasi dilakukan dengan mendefinisikan:

$$v = \frac{dx}{dt}$$

Sehingga diperoleh sistem:

$$\frac{dx}{dt} = v,\tag{2}$$

$$\frac{dv}{dt} = -\frac{k}{m}x - \frac{b}{m}v.\tag{3}$$

Kedua persamaan ini merupakan model yang digunakan dalam proses integrasi numerik untuk menentukan evolusi posisi dan kecepatan terhadap waktu.

---

## **Metode Runge–Kutta Orde Empat (RK4)**

Setelah model matematis diperoleh, langkah berikutnya adalah menentukan metode numerik yang digunakan untuk menyelesaikan Sistem (2)–(3). Pada proyek ini digunakan metode **Runge–Kutta orde empat (RK4)**, yang terkenal memiliki keseimbangan optimal antara akurasi dan efisiensi komputasi. RK4 sangat sesuai untuk sistem dinamis seperti osilasi teredam karena mampu menjaga stabilitas solusi dalam interval waktu yang cukup panjang.

Prinsip metode RK4 adalah mendekati solusi persamaan diferensial dengan menghitung empat nilai kemiringan (slope) dalam satu langkah waktu. Keempat kemiringan tersebut memberikan informasi mengenai perubahan sistem pada titik awal, dua titik tengah, dan titik akhir interval waktu. Untuk langkah waktu (h), tahap-tahap RK4 adalah sebagai berikut:

### **Tahap 1 (K₁): evaluasi pada titik awal**

$$K_{1x} = h , v_n,$$

$$K_{1v} = h\left(-\frac{k}{m}x_n - \frac{b}{m}v_n\right)$$

### **Tahap 2 (K₂): evaluasi pada titik tengah menggunakan K₁**

$$K_{2x} = h\left(v_n + \frac{1}{2}K_{1v}\right),$$

$$K_{2v} = h\left(-\frac{k}{m}\left(x_n + \frac{1}{2}K_{1x}\right) - \frac{b}{m}\left(v_n + \frac{1}{2}K_{1v}\right)\right)$$

### **Tahap 3 (K₃): evaluasi ulang titik tengah menggunakan K₂**

$$K_{3x} = h\left(v_n + \frac{1}{2}K_{2v}\right),$$

$$K_{3v} = h\left(-\frac{k}{m}\left(x_n + \frac{1}{2}K_{2x}\right) - \frac{b}{m}\left(v_n + \frac{1}{2}K_{2v}\right)\right)$$

### **Tahap 4 (K₄): evaluasi pada titik akhir interval**

$$K_{4x} = h\left(v_n + K_{3v}\right),$$

$$K_{4v} = h\left(-\frac{k}{m}(x_n + K_{3x}) - \frac{b}{m}(v_n + K_{3v})\right)$$

### **Pembaruan nilai posisi dan kecepatan**

$$x_{n+1} = x_n + \frac{1}{6}(K_{1x} + 2K_{2x} + 2K_{3x} + K_{4x}),$$

$$v_{n+1} = v_n + \frac{1}{6}(K_{1v} + 2K_{2v} + 2K_{3v} + K_{4v})$$

Dengan pemilihan langkah waktu (h) yang cukup kecil, metode RK4 mampu menghasilkan solusi numerik yang akurat dan stabil, sehingga perubahan amplitudo, redaman, serta dinamika osilasi dapat direpresentasikan dengan baik sepanjang simulasi.

# **C. ALGORITMA DENGAN MENGGUNAKAN PYTHON**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter fisika
m = 1.0          # massa
k = 20.0         # konstanta pegas
b = 1.5          # koefisien redaman

# Parameter waktu
h = 0.01         # langkah waktu
T = 20           # total waktu simulasi
N = int(T/h)     # jumlah iterasi

# Array waktu
t = np.linspace(0, T, N)

# Inisialisasi posisi dan kecepatan
x = np.zeros(N)
v = np.zeros(N)

# Kondisi awal
x[0] = 1.0       # posisi awal
v[0] = 0.0       # kecepatan awal

# Fungsi turunan dx/dt dan dv/dt
def f1(v):
    return v

def f2(x, v):
    return -(k/m)*x - (b/m)*v

# Implementasi metode Runge–Kutta orde 4
for i in range(N-1):

    k1_x = h * f1(v[i])
    k1_v = h * f2(x[i], v[i])

    k2_x = h * f1(v[i] + 0.5*k1_v)
    k2_v = h * f2(x[i] + 0.5*k1_x, v[i] + 0.5*k1_v)

    k3_x = h * f1(v[i] + 0.5*k2_v)
    k3_v = h * f2(x[i] + 0.5*k2_x, v[i] + 0.5*k2_v)

    k4_x = h * f1(v[i] + k3_v)
    k4_v = h * f2(x[i] + k3_x, v[i] + k3_v)

    x[i+1] = x[i] + (1/6) * (k1_x + 2*k2_x + 2*k3_x + k4_x)
    v[i+1] = v[i] + (1/6) * (k1_v + 2*k2_v + 2*k3_v + k4_v)

# Plot hasil
plt.figure(figsize=(10,5))
plt.plot(t, x, label="Posisi")
plt.xlabel("Waktu (s)")
plt.ylabel("Posisi (m)")
plt.title("Simulasi Osilasi Teredam Menggunakan Metode RK4")
plt.grid()
plt.legend()
plt.show()
```
