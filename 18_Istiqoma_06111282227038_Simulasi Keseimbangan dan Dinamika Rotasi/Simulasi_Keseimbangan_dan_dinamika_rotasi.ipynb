Dinamika rotasi adalah ilmu yang mempelajari tentang gerak rotasi (berputar dengan memperhatikan aspek penyebabnya, yaitu momen gaya. Momen gaya atau

# **1. Keseimbangan Statis Benda Tegar**

Dalam *sistem partikel*, benda dianggap sebagai suatu titik materi. Semua gaya yang bekerja pada benda dianggap bekerja pada titik materi tersebut, sehingga gaya yang bekerja dapat menyebabkan gerak translasi (tidak menyebabkan gerak rotasi).  
Oleh karena itu, syarat keseimbangan pada sistem partikel hanya berlaku untuk **keseimbangan translasi**:

$ ‚àëF = 0 $

## **Benda Tegar**
Benda tegar merupakan benda yang tidak berubah bentuk ketika diberi gaya tertentu.  
Berbeda dengan partikel, benda tegar memiliki banyak titik yang saling terhubung dan dapat menyebabkan gerak translasi dan rotasi.

Maka, syarat keseimbangan benda tegar adalah:

---

$ ‚àëF_{\text{y}} = 0 $


### **b. Tidak ada resultan momen gaya (torka)**

## **2. Dinamika Rotasi**

Dinamika rotasi adalah ilmu yang mempelajari tentang gerak rotasi (berputar)
dengan memperhatikan aspek penyebabnya, yaitu momen gaya. Momen gaya atau
yang lebih dikenal dengan torsi ini akan menyebabkan terjadinya percepatan
sudut. Suatu benda dikatakan melakukan gerak rotasi (berputar) jika semua bagian
benda bergerak mengelilingi poros atau sumbu putar. Sumbu putar benda terletak
pada salah satu bagian dari benda tersebut.

Benda tegar merupakan benda yang tidak mengalami perubahan bentuk akibat
pengaruh gaya, sehingga dalam melakukan pergerakan, benda tersebut tidak
mengalami perubahan bentuk dan volume benda. Benda tegar dapat melakukan gerak
translasi dan rotasi


##  **DINAMIKA ROTASI**

```python
import numpy as np
import matplotlib.pyplot as plt

# Kecepatan sudut konstan
omega = 2   # rad/s
t = np.linspace(0, 10, 200)  # 10 detik
theta = omega * t            # sudut rotasi

plt.figure(figsize=(8,4))
plt.plot(t, theta)
plt.xlabel('Waktu (s)')
plt.ylabel('Sudut (rad)')
plt.title('Rotasi dengan Kecepatan Sudut Konstan')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Percepatan sudut konstan
alpha = 0.5  # rad/s^2
t = np.linspace(0, 10, 200)

omega = alpha * t          # kecepatan sudut
theta = 0.5 * alpha * t**2 # sudut rotasi

plt.figure(figsize=(12,4))

# Grafik kecepatan sudut
plt.subplot(1,2,1)
plt.plot(t, omega)
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan Sudut (rad/s)')
plt.title('Grafik œâ(t) dengan Œ± konstan')
plt.grid()

# Grafik sudut
plt.subplot(1,2,2)
plt.plot(t, theta)
plt.xlabel('Waktu (s)')
plt.ylabel('Sudut (rad)')
plt.title('Grafik Œ∏(t) dengan Œ± konstan')
plt.grid()

plt.tight_layout()
plt.show()
```

Semakin besar momen inersia, garis semakin curam (torsi semakin besar untuk Œ± yang sama).

Persamaan dasar:

œÑ = I Œ±

```python
import numpy as np
import matplotlib.pyplot as plt

# Berbagai momen inersia (I)
I_values = [0.5, 1, 2, 3, 4]  # kg¬∑m^2

# Rentang percepatan sudut
alpha = np.linspace(0, 5, 100)

plt.figure(figsize=(8, 6))

for I in I_values:
    tau = I * alpha     # œÑ = I ¬∑ Œ±
    plt.plot(alpha, tau, label=f"I = {I} kg¬∑m¬≤")

plt.title("Grafik Hubungan Torsi (œÑ) dan Percepatan Sudut (Œ±) untuk Berbagai Momen Inersia")
plt.xlabel("Percepatan Sudut Œ± (rad/s¬≤)")
plt.ylabel("Torsi œÑ (N¬∑m)")
plt.grid(True)
plt.legend()
plt.show()
```

## **Kesimpulan**

1. Keseimbangan Benda Tegar

Untuk benda tidak bergerak dan stabil, gaya total dan momen total harus sama dengan nol. Prinsip ini digunakan pada palang, jembatan, timbangan, tumpuan balok, dan mekanika konstruksi.

2. Dinamika Rotasi

Gerak putar terjadi karena adanya torsi. Besarnya efek torsi tergantung pada: besar gaya, jarak gaya terhadap titik putar, momen inersia benda.

Rumus dasar: œÑ = I Œ± , yang menjelaskan hubungan antara torsi dan percepatan sudut.
