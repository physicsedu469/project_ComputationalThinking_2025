#**Simulasi Kecepatan dalam Gerak Lurus**

Kecepatan merupakan besaran yang menyatakan perubahan posisi benda terhadap waktu. Dalam gerak lurus, kecepatan dapat berupa konstan, bertambah, atau berkurang.

Secara umum, kecepatan didefinisikan:

$$
v = \frac{s}{t}
$$

Keterangan:


*   v = kecepatan (m/s)
*   s = jarak (m)
*   t = waktu (s)

Note:

*   Jika kecepatan tetap → GLB (Gerak Lurus Beraturan).
*   Jika kecepatan berubah teratur → GLBB (Gerak Lurus Berubah Beraturan).

##**1. Gerak Lurus dengan Kecepatan Konstan (GLB)**   

Pada gerak dengan kecepatan konstan, percepatan sama dengan nol. Persamaan posisi:

$$
s = v \cdot t
$$

Berikut simulasi gerak dengan kecepatan $v= 5 m/s$

```python
import numpy as np
import matplotlib.pyplot as plt

# Kecepatan konstan
v = 5   # m/s
t = np.linspace(0, 10, 200)  # 10 detik
s = v * t                    # posisi

plt.figure(figsize=(8,4))
plt.plot(t, s)
plt.xlabel('Waktu (s)')
plt.ylabel('Jarak (m)')
plt.title('Gerak Lurus dengan Kecepatan Konstan')
plt.grid(True)
plt.show()
```

Grafik jarak-waktu di atas membentuk garis lurus karena benda bergerak dengan kecepatan konstan. Kecepatan yang tetap membuat jarak bertambah secara linier terhadap waktu. Kemiringan garis menunjukkan besar kecepatan, yaitu semakin curam garisnya, semakin besar kecepatannya. Jika kecepatan diperbesar misalnya menjadi 10 m/s, garis akan semakin miring karena jarak bertambah lebih cepat dalam waktu yang sama.

##**2. Komponen Kecepatan dan Percepatan (GLBB)**

Apabila sebuah benda mengalami percepatan ($a$), maka:

$$
v (t) = v_0 + a t
$$

$$
s (t) = v_0t + \frac{1}{2} a t^2
$$

Dengan percepatan $a = 2 m/s^2$, berikut simulasi posisi dan kecepatan.

```python
a = 2  # percepatan
t = np.linspace(0, 10, 200)

v = a * t
s = 0.5 * a * t**2

plt.figure(figsize=(12,4))

# Grafik kecepatan
plt.subplot(1,2,1)
plt.plot(t, v)
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Grafik Kecepatan terhadap Waktu v(t)')
plt.grid()

# Grafik jarak
plt.subplot(1,2,2)
plt.plot(t, s)
plt.xlabel('Waktu (s)')
plt.ylabel('Jarak (m)')
plt.title('Grafik Jarak terhadap Waktu s(t)')
plt.grid()

plt.tight_layout()
plt.show()
```

Grafik kecepatan-waktu berupa garis lurus karena kecepatan bertambah secara konstan tiap detik. Sementara itu, grafik jarak-waktu berbentuk kurva karena jarak bertambah semakin cepat sesuai percepatan.

##**3. Membandingkan Kecepatan Konstan dan Kecepatan Berubah**


```python
v_konstan = np.full_like(t, 5)  # 5 m/s

plt.figure(figsize=(8,4))
plt.plot(t, v_konstan, label='Kecepatan Konstan 5 m/s')
plt.plot(t, v, label='Kecepatan dengan Percepatan 2 m/s²')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Perbandingan Kecepatan Konstan dan Bertambah')
plt.legend()
plt.grid(True)
plt.show()
```

Garis biru menunjukkan kecepatan yang tetap sebesar 5 m/s, sehingga grafiknya datar. Garis oranye menunjukkan kecepatan yang terus meningkat karena adanya percepatan. Kedua garis berpotongan pada satu titik, yaitu saat kecepatan akibat percepatan menyamai kecepatan konstan. Setelah itu, kecepatan pada garis oranye menjadi lebih besar. Percepatan terjadi karena ada gaya yang mendorong benda, sedangkan kecepatan konstan berarti tidak ada gaya tambahan yang membuatnya berubah.

##**4. Analisis Percepatan dalam GLBB**

Percepatan konstan digambarkan sebagai garis mendatar karena nilainya tidak berubah terhadap waktu.

```python
plt.figure(figsize=(8,3))
plt.plot(t, np.full_like(t, a))
plt.xlabel('Waktu (s)')
plt.ylabel('Percepatan (m/s²)')
plt.title('Grafik Percepatan pada GLBB')
plt.grid(True)
plt.show()
```

Percepatan konstan berarti perubahan kecepatan per satuan waktu selalu sama. Jika percepatan bernilai negatif, grafik akan berada di bawah sumbu horizontal dan benda mengalami perlambatan (kecepatan berkurang).
