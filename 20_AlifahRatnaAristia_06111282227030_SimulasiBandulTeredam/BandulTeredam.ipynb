#**Simulasi Bandul Teredam dengan Hambatan Udara**

Sebuah bandul sederhana bergerak karena adanya gaya gravitasi sebagai gaya pemulih.
Namun pada keadaan nyata, udara memberikan gaya hambatan yang selalu berlawanan dengan arah gerak bandul.
Gaya hambatan ini menyebabkan energi sistem berkurang sehingga osilasinya semakin melemah (teredam).


Gaya hambatan dimodelkan sebagai:

$$
\vec{F_d} = -b\vec{v}
$$


dengan b adalah konstanta redaman udara (damping coefficient).
Maka persamaan geraknya berdasarkan Hukum II Newton:

$$
m\vec{a} = \vec{F_g} + \vec{F_d}
$$

Untuk gerak bandul pada sudut kecil, berlaku persamaan:

$$
\theta'' + \frac{b}{m}\theta' + \frac{g}{L}\sin(\theta) = 0
$$

Sehingga percepatan sudut bandul menjadi:

$$
\alpha(t) = -\frac{b}{m}\omega(t) - \frac{g}{L}\sin(\theta(t))
$$

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter fisika
g = 9.81
L = 1.0
m = 1.0
b = 0.25
```

```python
# Setup waktu
dt = 0.01
t_max = 20
t = np.arange(0, t_max, dt)
```

```python
# Variabel sudut dan kecepatan sudut
theta = np.zeros_like(t)
omega = np.zeros_like(t)
```

```python
theta[0] = np.radians(30)
omega[0] = 0
```

```python
# Simulasi Euler
for i in range(len(t)-1):
    alpha = -(b/m)*omega[i] - (g/L)*np.sin(theta[i])
    omega[i+1] = omega[i] + alpha*dt
    theta[i+1] = theta[i] + omega[i+1]*dt
```

```python
# Energi mekanik
E = 0.5*m*(L**2)*(omega**2) + m*g*L*(1 - np.cos(theta))
```

```python
# Grafik 1 - Sudut
plt.figure(figsize=(9,4))
plt.plot(t, np.degrees(theta))
plt.title("Sudut Bandul terhadap Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Sudut (derajat)")
plt.grid()
plt.show()
```

```python
# Grafik 2 - Energi Mekanik
plt.figure(figsize=(9,4))
plt.plot(t, E, color='pink')
plt.title("Energi Mekanik Bandul terhadap Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Energi (Joule)")
plt.grid()
plt.show()
```

## **Analisis Grafik**

 Grafik sudut terhadap waktu menunjukkan bahwa amplitudo osilasi semakin mengecil.
Hal ini disebabkan oleh gaya hambatan udara yang menyebabkan energi berkurang.

 Grafik energi mekanik terhadap waktu menurun secara eksponensial.
Energi berubah menjadi energi panas karena adanya interaksi udara.

 Kesimpulan: Bandul teredam merupakan sistem tidak konservatif karena energi mekanik tidak konstan.

Simulasi ini telah sesuai dengan konsep fisika pada bandul nyata.
