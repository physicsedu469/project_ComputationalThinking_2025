#**Simulasi Rangkaian Listrik dengan Hukum Ohm**

Sebuah rangkaian listrik sederhana terdiri dari sumber tegangan $V$, hambatan listrik $R$, dan arus listrik $I$ yang mengalir dalam rangkaian. Pada rangkaian ideal, hubungan antara tegangan, arus, dan hambatan dapat dijelaskan menggunakan Hukum Ohm. Hukum ini menyatakan bahwa besar arus listrik yang mengalir sebanding dengan tegangan yang diberikan dan berbanding terbalik dengan hambatan rangkaian.

Secara matematis, Hukum Ohm dinyatakan sebagai:

$V = I . R$
_____________________________________

Dimana:

$V$ adalah tegangan listrik (volt)

$I$ adalah arus listrik (Ampere)

$R$ adalah hambatan listrik (ohm)
_____________________________________

Jika nilai tegangan dan hambatan diketahui, maka arus listrik yang mengalir dalam rangkaian dapat dihitung menggunakan persamaan:

$I = \frac{V}{R}$

Sebaliknya, jika arus dan hambatan diketahui, maka tegangan dapat ditentukan melalui:

$V = I . R$

Hubungan ini menjadi dasar dalam simulasi rangkaian listrik untuk menganalisis pengaruh perubahan tegangan atau hambatan terhadap besar arus listrik yang mengalir. Dengan memvariasikan salah satu parameter, perilaku rangkaian dapat diamati secara kuantitatif sesuai prinsip Hukum Ohm.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def current_update(I, V, R, dt):
    dI = (V / R - I) * dt
    I_new = I + dI
    return I_new, dI

def charge_update(Q, I, dt):
    return Q + I * dt

def ohm_simulation(V, R, dt=0.01, t_max=5):
    I = 0.0
    Q = 0.0
    t = 0.0

    Is, Qs, dIs, ts = [I], [Q], [], [t]

    while t <= t_max:
        I, dI = current_update(I, V, R, dt)
        Q = charge_update(Q, I, dt)
        t += dt

        Is.append(I)
        Qs.append(Q)
        dIs.append(dI)
        ts.append(t)

    return Is, Qs, dIs, ts
```

**Hubungan Tegangan, Arus dan Hambatan**

Grafik yang akan dibuat menunjukkan hubungan antara tegangan listrik, arus listrik dan Hambatan listrik pada suatu rangkaian berdasarkan Hukum Ohm. Hubungan ini dinyatakan oleh persamaan:

$V = I . R$

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def voltage(I, R):
    return I * R

def voltage_vs_current(R, I_max=10.0, dI=0.1):
    I = 0.0
    V = voltage(I, R)

    Is, Vs = [I], [V]

    while I <= I_max:
        I += dI
        V = voltage(I, R)
        Is.append(I)
        Vs.append(V)

    return Is, Vs

R = 5.0
I_max = 10.0
dI = 0.1

I_vals, V_vals = voltage_vs_current(R, I_max, dI)

plt.figure(figsize=(8,6))
plt.plot(I_vals, V_vals, label=f'R = {R} Ω')
plt.xlabel('Arus I (A)')
plt.ylabel('Tegangan V (V)')
plt.title('Hubungan Tegangan terhadap Arus (V = I · R)')
plt.legend()
plt.grid(True)
plt.show()
```

**Hubungan Hambatan, Tegangan dan Arus**

Grafik yang akan dibuat menunjukkan hubungan antara Hambatan listrik, Tegangan listrik, dan Arus listrik pada suatu rangkaian berdasarkan Hukum Ohm. Hubungan ini dinyatakan oleh persamaan:

$R = \frac{V}{I}$

```python
import numpy as np
import matplotlib.pyplot as plt

def resistance(V, I):
    return V / I

def resistance_vs_voltage(I, V_max=20.0, dV=0.5):
    V = dV
    R = resistance(V, I)

    Vs, Rs = [V], [R]

    while V <= V_max:
        V += dV
        R = resistance(V, I)
        Vs.append(V)
        Rs.append(R)

    return Vs, Rs

I1 = 2.0
I2 = 4.0
V_max = 20.0
dV = 0.5

V_I1, R_I1 = resistance_vs_voltage(I1, V_max, dV)
V_I2, R_I2 = resistance_vs_voltage(I2, V_max, dV)

plt.figure(figsize=(8,6))
plt.plot(V_I1, R_I1, label=f'I = {I1} A')
plt.plot(V_I2, R_I2, label=f'I = {I2} A')
plt.xlabel('Tegangan V (V)')
plt.ylabel('Hambatan R (Ω)')
plt.title('Hubungan Hambatan terhadap Tegangan (R = V / I)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def resistance(V, I):
    return V / I

def resistance_vs_current(V, I_max=10.0, dI=0.5):
    I = dI
    R = resistance(V, I)

    Is, Rs = [I], [R]

    while I <= I_max:
        I += dI
        R = resistance(V, I)
        Is.append(I)
        Rs.append(R)

    return Is, Rs

V1 = 2
V2 = 4

I_max = 10.0
dI = 0.5

I_V1, R_V1 = resistance_vs_current(V1, I_max, dI)
I_V2, R_V2 = resistance_vs_current(V2, I_max, dI)

plt.figure(figsize=(8,6))
plt.plot(I_V1, R_V1, label=f'V = {V1} V')
plt.plot(I_V2, R_V2, label=f'V = {V2} V')
plt.xlabel('Arus I (A)')
plt.ylabel('Hambatan R (Ω)')
plt.title('Hubungan Hambatan terhadap Arus (R = V / I)')
plt.legend()
plt.grid(True)
plt.show()
```

**Penjelasan Grafik Hambatan Terhadap Tegangan (Tegangan Konstan)**

Grafik kedua memperlihatkan hubungan antara hambatan dan arus listrik pada rangkaian dengan tegangan yang dijaga tetap. Grafik dengan tegangan yang lebih besar menghasilkan nilai hambatan yang lebih tinggi pada setiap arus dibandingkan grafik dengan tegangan yang lebih kecil.

**Penjelasan Grafik Hambatan terhadap Tegangan (Arus Konstan)**

Grafik pertama menunjukkan hubungan antara hambatan listrik dan tegangan pada rangkaian dengan nilai arus yang dijaga konstan. Terlihat bahwa hambatan meningkat secara linier seiring bertambahnya tegangan untuk masing-masing nilai arus. Sesuai dengan persamaan Hukum Ohm $R = \frac{V}{I}$ , gradien grafik ditentukan oleh nilai kebalikan arus. Arus yang lebih kecil menghasilkan kemiringan grafik yang lebih besar, sehingga untuk tegangan yang sama diperoleh nilai hambatan yang lebih besar dibandingkan arus yang lebih besar.

**Perbandingan arus listrik pada rangkaian dengan hambatan berbeda $V = 12V$**

$R = 5Ω$ , atau

$R = 10Ω$

```python
I_lowR, Q_lowR, dI_lowR, t_lowR = ohm_simulation(V=12, R=5)
I_highR, Q_highR, dI_highR, t_highR = ohm_simulation(V=12, R=10)

plt.figure(figsize=(10, 12))

plt.subplot(3,1,1)
plt.plot(t_lowR, I_lowR, label='hambatan kecil (R = 5 Ω)')
plt.plot(t_highR, I_highR, label='hambatan besar (R = 10 Ω)')
plt.xlabel('waktu (s)')
plt.ylabel('arus (A)')
plt.title('Perbandingan Arus Listrik terhadap Waktu')
plt.legend()
```

**Penjelasan :**

Grafik pertama menunjukkan perubahan arus listrik terhadap waktu pada dua rangkaian dengan nilai hambatan yang berbeda. Terlihat bahwa rangkaian dengan hambatan lebih kecil menghasilkan arus listrik yang lebih besar dibandingkan rangkaian dengan hambatan yang lebih besar pada kondisi sumber yang sama. Hal ini sesuai dengan Hukum Ohm yang dinyatakan oleh persamaan

$I = \frac{V}{R}$

yang menunjukkan bahwa besar arus listrik berbanding terbalik dengan hambatan rangkaian. Grafik ini juga memperlihatkan bahwa arus pada rangkaian dengan hambatan kecil mencapai kondisi stabil lebih cepat dibandingkan rangkaian dengan hambatan yang lebih besar.

Berdasarkan perbedaan perilaku arus listrik tersebut, diperlukan analisis lanjutan untuk melihat bagaimana hubungan antara arus yang dihasilkan dengan tegangan listrik pada masing-masing hambatan. Oleh karena itu, pada grafik berikutnya akan ditampilkan perubahan tegangan listrik terhadap waktu yang dihitung berdasarkan nilai arus pada grafik pertama menggunakan hubungan Hukum Ohm.

```python
V_lowR = [I * 5 for I in I_lowR]
V_highR = [I * 10 for I in I_highR]

plt.plot(t_lowR, V_lowR, label='tegangan (R = 5 Ω)')
plt.plot(t_highR, V_highR, label='tegangan (R = 10 Ω)')

plt.xlabel('Waktu (s)')
plt.ylabel('Tegangan (V)')
plt.title('Perubahan Tegangan terhadap Waktu')
plt.legend()
plt.grid(True)
```

#**Kesimpulan**

Simulasi rangkaian listrik ini menunjukkan bahwa Hukum Ohm berlaku secara konsisten dalam berbagai kondisi rangkaian. Grafik-grafik yang dihasilkan menggambarkan hubungan langsung antara tegangan, arus, dan hambatan, di mana perubahan pada salah satu variabel akan memengaruhi dua variabel lainnya sesuai persamaan dasar ($V = I . R$). Ketika hambatan dijaga tetap, peningkatan tegangan terbukti menghasilkan arus yang meningkat secara linear. Hal ini menegaskan bahwa pada hambatan konstan, arus selalu berbanding lurus dengan tegangan.

Di sisi lain, ketika tegangan dijaga tetap, perubahan hambatan memberikan dampak yang signifikan terhadap besar arus. Hambatan yang lebih besar menghasilkan arus yang lebih kecil, sedangkan hambatan yang lebih kecil memungkinkan arus mengalir lebih besar. Perbandingan grafik antara hambatan 5Ω dan 10Ω memperjelas pola ini. Dengan demikian, simulasi ini tidak hanya memvalidasi prinsip dasar Hukum Ohm, tetapi juga memberikan pemahaman visual yang kuat mengenai bagaimana tegangan, arus, dan hambatan saling berkaitan dalam suatu rangkaian listrik.
