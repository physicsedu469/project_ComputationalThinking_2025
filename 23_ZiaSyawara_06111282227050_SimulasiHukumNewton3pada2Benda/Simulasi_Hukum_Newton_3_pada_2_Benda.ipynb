#**Simulasi Hukum Newton 3 Pada 2 Benda**

Hukum Newton 3 menyatakan bahwa setiap aksi akan menimbulkan reaksi yang sama besar namun berlawanan arah. Dalam konteks fisika, ketika dua benda berinteraksi, keduanya saling memberikan gaya yang besarnya sama tetapi arahnya berlawanan.
___
Misalnya terdapat dua benda, A dan B, yang bersentuhan dan saling mendorong. Gaya aksi-reaksi dapat dituliskan sebagaiberikut:
$\vec{F}_{A \rightarrow B} = -\vec{F}_{B \rightarrow A}$

dengan $\vec{F}_{A \rightarrow B}$ adalah gaya yang diberikan benda A pada benda B, dan $\vec{F}_{B \rightarrow A}$ adalah gaya yang diberikan benda B pada benda A
___
Menurut Hukum Newton 2, Masing- masing dari benda tersebut akan memiliki percepatan akibat dari gaya total yang bekerja pada benda tersebut. Maka persamaan geraknya akan menjadi:
$m_A \vec{a}_A = \vec{F}_{B \rightarrow A}$

$m_B \vec{a}_B = \vec{F}_{A \rightarrow B}$
___
karena setiap aksi akan ada reaksi maka:
$\vec{F}_{B \rightarrow A} = -\vec{F}_{A \rightarrow B}$
Sehingga Percepatan dari masing-masing benda dapat ditulis menjadi:
$\vec{a}_A = \frac{\vec{F}}{m_A},
\qquad
\vec{a}_B = -\frac{\vec{F}}{m_B}$

dengan $\vec{F}$ adalah besar gaya interaksi antara kedua benda.

```python
import numpy as np
import matplotlib.pyplot as plt

def hitung_percepatan(F, mA, mB):
    aA = F / mA
    aB = -F / mB
    return aA, aB

def update_kecepatan(vA, vB, aA, aB, dt):
    vA_baru = vA + aA * dt
    vB_baru = vB + aB * dt
    return vA_baru, vB_baru

def update_posisi(xA, xB, vA, vB, dt):
    xA_baru = xA + vA * dt
    xB_baru = xB + vB * dt
    return xA_baru, xB_baru

def simulasi_newton3(F, mA, mB, vA_awal=0, vB_awal=0, dt=0.01, t_maks=5):
    vA, vB = vA_awal, vB_awal
    xA, xB = 0.0, 0.0
    t = 0.0

    daftar_xA, daftar_xB = [xA], [xB]
    daftar_vA, daftar_vB = [vA], [vB]
    daftar_aA, daftar_aB = [], []
    daftar_t = [t]

    while t <= t_maks:
        aA, aB = hitung_percepatan(F, mA, mB)
        vA, vB = update_kecepatan(vA, vB, aA, aB, dt)
        xA, xB = update_posisi(xA, xB, vA, vB, dt)
        t += dt

        daftar_xA.append(xA)
        daftar_xB.append(xB)
        daftar_vA.append(vA)
        daftar_vB.append(vB)
        daftar_aA.append(aA)
        daftar_aB.append(aB)
        daftar_t.append(t)

    return daftar_xA, daftar_xB, daftar_vA, daftar_vB, daftar_aA, daftar_aB, daftar_t
```

Perbandingan Gerak dua benda dengan massa yang berbeda pada hukum newton 3

Massa A ringan, Massa B berat:

$mA$ = 1 kg

$mB$ = 5 kg

Gaya aksi-reaksi: $F$ = 10N

Massa A Berat, Massa B ringan :

$mA$ = 5 kg

$mB$ = 1 kg

```python
xA_ringan, xB_berat, vA1, vB1, aA1, aB1, t1 = simulasi_newton3(F=10, mA=1, mB=5)
xA_berat, xB_ringan, vA2, vB2, aA2, aB2, t2 = simulasi_newton3(F=10, mA=5, mB=1)

plt.figure(figsize=(10, 12))

plt.subplot(3,1,1)
plt.plot(t1, xA_ringan, label='mA ringan, mB berat (posisi A)')
plt.plot(t1, xB_berat, label='mA ringan, mB berat (posisi B)')
plt.plot(t2, xA_berat, label='mA berat, mB ringan (posisi A)')
plt.plot(t2, xB_ringan, label='mA berat, mB ringan (posisi B)')
plt.xlabel('Waktu (s)')
plt.ylabel('Posisi (m)')
plt.title('Perbandingan Posisi Dua Benda pada Hukum Newton 3')
plt.legend()
```

Pada grafik perbandingan posisi dua benda berdasarkan Hukum Newton 3, terlihat bahwa kedua benda bergerak saling menjauh karena gaya aksi-reaksi bekerja dengan besar sama namun arah berlawanan. Perbedaannya muncul karena massa kedua benda tidak sama: benda bermassa lebih kecil berpindah lebih cepat, sedangkan benda bermassa besar bergerak lebih lambat. Ini menunjukkan bahwa besar respon gerak dipengaruhi oleh massa, bukan hanya oleh gaya.

Hal ini dapat dijelaskan dengan Grafik perbandingan kecepatan yang juga menegaskan hal tersebut. Benda bermassa kecil mengalami peningkatan kecepatan lebih cepat karena percepatannya lebih besar, sementara benda bermassa besar mengalami perubahan kecepatan lebih lambat. Hal ini memperlihatkan bahwa meskipun gaya aksi-reaksi sama besar, efek geraknya tidak simetris akibat perbedaan massa masing-masing benda.

```python
plt.plot(t1, vA1, label='v_A (mA kecil, mB besar)')
plt.plot(t1, vB1, label='v_B (mA kecil, mB besar)')
plt.plot(t2, vA2, '--', label='v_A (mA besar, mB kecil)')
plt.plot(t2, vB2, '--', label='v_B (mA besar, mB kecil)')
plt.xlabel('Waktu (s)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Perbandingan Kecepatan pada Dua Massa yang Berbeda')
plt.legend()
```

Perbandingan kecepatan kedua benda memperlihatkan bahwa meskipun keduanya saling berinteraksi, perubahan kecepatannya tidak sama. Benda yang massanya lebih kecil mengalami perubahan kecepatan yang jauh lebih besar, sedangkan benda bermassa lebih besar berubah lebih lambat. Perbedaan respons ini menuntut pembahasan lebih lanjut mengenai sumber utama perubahannya, yaitu gaya yang bekerja pada masing-masing benda.

Untuk itu, maka grafik gaya aksi-reaksi ditampilkan guna menunjukkan bahwa gaya interaksi di antara kedua benda ternyata selalu sama besar dan berlawanan arah. Dengan kata lain, perbedaan pola kecepatan sebelumnya bukan disebabkan oleh gaya yang berbeda, melainkan oleh perbedaan massa masing-masing benda. Hubungan ini menegaskan bahwa percepatan dan kecepatan yang terlihat pada grafik sebelumnya sepenuhnya merupakan konsekuensi dari kombinasi Hukum Newton II dan Hukum Newton III yang bekerja secara bersamaan dalam sistem dua benda.

```python
xA, xB, vA, vB, aA, aB, t = simulasi_newton3(F=10, mA=5, mB=1)
F = 10
F_A_ke_B = np.full(len(t), F)
F_B_ke_A = np.full(len(t), -F)
plt.figure(figsize=(8,5))
plt.plot(t, F_A_ke_B, label='F_A  B (Aksi)')
plt.plot(t, F_B_ke_A, label='F_B  A (Reaksi)')
plt.xlabel('Waktu (s)')
plt.ylabel('Gaya (N)')
plt.title('Grafik Gaya AksiReaksi (Hukum Newton 3)')
plt.grid(True)
plt.legend()
plt.show()
```

Pada grafik gaya aksireaksi terlihat bahwa gaya yang bekerja pada kedua benda selalu memiliki besar yang sama tetapi arah yang berlawanan. Hal ini sesuai dengan Hukum Newton 3 yang menyatakan bahwa setiap aksi selalu memiliki reaksi yang besarnya sama dan arahnya berlawanan. Meskipun kedua benda memiliki massa yang berbeda, gaya yang mereka berikan satu sama lain tetap konstan dan simetris sepanjang waktu.

Perbedaan massa tidak memengaruhi besar gaya, tetapi berpengaruh pada percepatan masing-masing benda. Hal ini kemudian tercermin pada grafik jarak antar benda. Ketika benda dengan massa lebih kecil mengalami percepatan yang lebih besar, jarak antara kedua benda akan berubah lebih cepat dibandingkan benda dengan massa yang lebih besar. Dengan demikian, kombinasi kedua grafik ini menunjukkan bahwa meskipun gaya aksireaksi selalu sama besar, respons gerak kedua benda bisa sangat berbeda bergantung pada massanya.

```python
xA, xB, vA, vB, aA, aB, t = simulasi_newton3(F=10, mA=5, mB=1)

jarak = np.abs(np.array(xA) - np.array(xB))

plt.figure(figsize=(8,5))
plt.plot(t, jarak, label='Jarak |xA - xB|')
plt.xlabel('Waktu (s)')
plt.ylabel('Jarak Antar Benda (m)')
plt.title('Grafik Jarak Antar Dua Benda Terhadap Waktu')
plt.grid(True)
plt.legend()
plt.show()
```

Grafik jarak antara dua benda terhadap waktu memperlihatkan bagaimana kedua benda bergerak saling menjauh setelah gaya aksi-reaksi bekerja. Karena kedua benda memiliki massa yang berbeda, percepatan yang dialami masing-masing juga berbeda. Benda dengan massa lebih kecil mengalami percepatan yang lebih besar sehingga bergerak lebih cepat, menyebabkan jarak antara kedua benda bertambah secara bertahap seiring berjalannya waktu.

Perubahan jarak ini menunjukkan bahwa meskipun gaya yang bekerja pada kedua benda memiliki besar yang sama dan arah berlawanan, respons gerak tiap benda tidak sama. Massa menentukan seberapa besar percepatan yang terjadi, sehingga benda yang lebih ringan bergerak lebih jauh dalam waktu yang sama. Dengan demikian, grafik ini menegaskan hubungan antara gaya, massa, dan percepatan.
