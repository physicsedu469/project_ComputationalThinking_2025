```python
# Simulasi Pemuaian Linear

import numpy as np
import matplotlib.pyplot as plt
```

```python
def hitung_panjang(L0, alpha, T_awal, T_akhir):
    # L0      : panjang awal
    # alpha   : koefisien muai
    # T_awal  : suhu awal
    # T_akhir : suhu akhir

    deltaT = T_akhir - T_awal
    deltaL = alpha * L0 * deltaT
    L_akhir = L0 + deltaL

    return L_akhir
```

*L0 adalah panjang awal →* wajib untuk menghitung perubahan panjang.

*alpha adalah koefisien muai →* muncul langsung dari rumus fisika.

*T0 dan Takhir* menentukan rentang perubahan suhu yang ingin diuji.

*dt* menentukan besar langkah perubahan suhu.

```python
# Nilai awal
L0 = 1.0          # meter
alpha = 0.000012  # baja
T_awal = 25       # derajat C
T_akhir = 150     # derajat C
```

```python
# Simulasi kenaikan suhu bertahap
T = T_awal
kenaikan = 5     # supaya terlihat jelas perubahannya

daftar_suhu = []
daftar_panjang = []

while T <= T_akhir:
    L = hitung_panjang(L0, alpha, T_awal, T)

    daftar_suhu.append(T)
    daftar_panjang.append(L)

    T = T + kenaikan  # naik suhu sedikit demi sedikit
```

```python
print("HASIL PEMUAIAAN LINEAR BAJA")
print("-----------------------------")

for i in range(len(daftar_suhu)):
    print("Suhu:", daftar_suhu[i], "°C | Panjang:", daftar_panjang[i], "m")
```

```python
plt.plot(daftar_suhu, daftar_panjang, marker='o')
plt.title("Grafik Pemuaian Linear Baja")
plt.xlabel("Suhu (°C)")
plt.ylabel("Panjang (meter)")
plt.grid(True)
plt.show()
```

```python
alpha_baja = 0.000012
alpha_alum = 0.000023
alpha_tembaga = 0.000017

L_baja = []
L_alum = []
L_tembaga = []

for T in daftar_suhu:
    L_b = hitung_panjang(L0, alpha_baja, T_awal, T)
    L_a = hitung_panjang(L0, alpha_alum, T_awal, T)
    L_t = hitung_panjang(L0, alpha_tembaga, T_awal, T)

    L_baja.append(L_b)
    L_alum.append(L_a)
    L_tembaga.append(L_t)

plt.plot(daftar_suhu, L_baja, label="Baja")
plt.plot(daftar_suhu, L_alum, label="Aluminium")
plt.plot(daftar_suhu, L_tembaga, label="Tembaga")

plt.title("Perbandingan Pemuaian pada 3 Material")
plt.xlabel("Suhu (°C)")
plt.ylabel("Panjang (meter)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L0 = 1.0
T0 = 20
Takhir = 200
dt = 10
```

```python
materials = [
    ("Baja", 0.000012),
    ("Aluminium", 0.000023),
    ("Tembaga", 0.000017),
    ("Kaca", 0.000009)
]

T_list = list(range(T0, Takhir+1, dt))
```

```python
hasil = {}
for nama, alpha in materials:
    Ls = []
    for T in T_list:
        L = L0 * (1 + alpha*(T - T0))
        Ls.append(L)
    hasil[nama] = Ls
```

```python
print("Suhu | Panjang_Aluminium | % perubahan")
for i, T in enumerate(T_list):
    L = hasil["Aluminium"][i]
    persen = (L - L0)/L0 * 100
    print(f"{T:3d}°C | {L:.6f} m | {persen:.4f}%")
```

```python
for nama in hasil:
    plt.plot(T_list, hasil[nama], label=nama)
plt.xlabel("Suhu (°C)")
plt.ylabel("Panjang (m)")
plt.title("Perbandingan Pemuaian Material")
plt.legend()
plt.grid(True)
plt.show()
```
