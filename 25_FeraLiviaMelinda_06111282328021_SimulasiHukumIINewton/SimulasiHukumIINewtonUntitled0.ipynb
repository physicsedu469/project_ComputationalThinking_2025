# HUKUM II NEWTON

# Simulasi Komputasi Hukum II Newton

##1. Pendahuluan

Hukum II Newton merupakan salah satu hukum dasar dalam mekanika klasik yang menjelaskan hubungan antara gaya, massa, dan percepatan suatu benda. Hukum ini menyatakan bahwa percepatan suatu benda berbanding lurus dengan gaya total yang bekerja padanya serta berbanding terbalik dengan massanya.
Secara matematis, hukum II Newton dapat dituliskan sebagai:
        F=ma
dengan:
F = gaya total yang bekerja (Newton)
m = massa benda (kg)
a = percepatan (m/s)
Sehingga percepatan dapat dihitung:

          a=f/m

	


2. Metode Simulasi

Simulasi dilakukan menggunakan pendekatan numerik time-stepping, yaitu membagi waktu menjadi interval kecil (t), kemudian menghitung perubahan kecepatan dan posisi pada setiap langkah.



```python
import numpy as np
import matplotlib.pyplot as plt
```

Parameter

```python
# Parameter dasar simulasi
m = 2.0        # massa benda (kg)
F = 10.0       # gaya total (N)
t_max = 10.0   # waktu simulasi (detik)
dt = 0.01      # langkah waktu (s)
```

```python
# Hitung percepatan berdasarkan Hukum II Newton
a = F / m

print(f"Massa (m)     : {m} kg")
print(f"Gaya (F)      : {F} N")
print(f"Percepatan (a): {a:.2f} m/s")
```

# Perhitungan Numerik (Time-Stepping)

```python
# Membuat array waktu
t = np.arange(0, t_max + dt, dt)
```

```python
# Menyediakan array kecepatan dan posisi
v = np.zeros_like(t)
x = np.zeros_like(t)
```

```python
# Simulasi menggunakan metode time-stepping
for i in range(1, len(t)):
    v[i] = v[i-1] + a * dt          # v(t) = v(t-dt) + adt
    x[i] = x[i-1] + v[i] * dt       # x(t) = x(t-dt) + vdt
```

## Visualisasi Grafik

Grafik Percepatan

```python
plt.figure(figsize=(16, 5))

# Grafik percepatan
plt.subplot(1, 3, 1)
plt.plot(t, np.full_like(t, a))
plt.title("Percepatan vs Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Percepatan (m/s)")
plt.grid(True)
```

Grafik kecepatan

```python
plt.figure(figsize=(16, 5))

# Grafik kecepatan
plt.subplot(1, 3, 2)
plt.plot(t, v)
plt.title("Kecepatan vs Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Kecepatan (m/s)")
plt.grid(True)
```

Grafik posisi

```python
plt.figure(figsize=(16, 5))

# Grafik posisi
plt.subplot(1, 3, 3)
plt.plot(t, x)
plt.title("Posisi vs Waktu")
plt.xlabel("Waktu (s)")
plt.ylabel("Posisi (m)")
plt.grid(True)
```

##Simulasi Interaktif

```python
from ipywidgets import interact

def simulasi(F=10.0, m=2.0):
    a = F / m
    t = np.linspace(0, 10, 1000)
    v = a * t
    x = 0.5 * a * t**2

    plt.figure(figsize=(12,4))
    plt.plot(t, v, label="Kecepatan (m/s)")
    plt.plot(t, x, label="Posisi (m)")
    plt.title(f"Simulasi Hukum II Newton | F={F} N, m={m} kg, a={a:.2f} m/s")
    plt.xlabel("Waktu (s)")
    plt.ylabel("Nilai")
    plt.legend()
    plt.grid(True)
    plt.show()

interact(simulasi, F=(1, 50), m=(1, 20))
```

Menurut Hukum II Newton, gaya 10.0 N yang bekerja pada massa 2.0 kg menghasilkan percepatan 5.00 m/s.
Simulasi menunjukkan bahwa:
- Percepatan tetap (konstan).
- Kecepatan bertambah secara linear.
- Posisi bertambah secara kuadratis terhadap waktu.
