#**Gerak Lurus Beraturan**

**Pengertian GLB**

Gerak Lurus Beraturan adalah gerak suatu benda pada lintasan lurus dengan kecepatan konstan, sehingga jarak yang ditempuh tiap selang waktu selalu sama.

**Ciri-ciri GLB:**

- Lintasan lurus

- Kecepatan tetap (tidak berubah)

- Percepatan nol $(a = 0)$

**Persamaan Umum:**

$ùë•=ùë•0+ùë£ùë°$
_____________________________________
Dimana:

$x$ adalah posisi benda pada waktu t

$x‚ÇÄ$ adalah posisi awal (saat t = 0)

$v$ adalah kecepatan konstan

$t$ adalah waktu yang berjalan
_____________________________________

Rumus ini menunjukkan bahwa posisi benda pada waktu $t (x)$ merupakan hasil dari posisi awal $(x‚ÇÄ)$ ditambah jarak yang ditempuh selama bergerak, yaitu kecepatan $√ó$ waktu $(vt)$. Karena v tetap, maka pertambahan jarak selalu linear terhadap waktu.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def velocity_update(v, dt):
    dv = 0.0
    return v, dv

def position_update(x, v, dt):
    dx = v * dt
    x_new = x + dx
    return x_new, dx

def acceleration_update():
    return 0.0

def glb_simulation(v, x0=0.0, dt=0.1, t_max=10.0):
    x = x0
    t = 0.0
    a = acceleration_update()

    xs = [x]
    vs = [v]
    accs = [a]
    ts = [t]


    while t + dt <= t_max + 1e-12:
        v, dv = velocity_update(v, dt)
        x, dx = position_update(x, v, dt)
        a = acceleration_update()
        t += dt

        xs.append(x)
        vs.append(v)
        accs.append(a)
        ts.append(t)

    return xs, vs, accs, ts
```

**Hubungan antara Posisi dan Waktu $(x‚Äìt)$**

Pada Gerak Lurus Beraturan (GLB), posisi benda berubah secara linear terhadap waktu karena kecepatan bernilai konstan. Hal ini menyebabkan grafik posisi‚Äìwaktu berbentuk garis lurus miring.

**Persamaan hubungan posisi‚Äìwaktu:**

$x(t)=x0‚Äã+vt$

```python
import math
import numpy as np
import matplotlib.pyplot as plt

v = 5.0    # m/s
x0 = 2.0   # m
dt = 0.1
t_max = 8.0

ts = np.arange(0, t_max+dt, dt)
xs = x0 + v * ts

plt.figure(figsize=(6,4))
plt.plot(ts, xs, linewidth=2)
plt.scatter([0,1,2], [x0 + v*ti for ti in (0,1,2)], zorder=5)
plt.title("Posisi vs Waktu (GLB)")
plt.xlabel("Waktu (s)")
plt.ylabel("Posisi x (m)")
plt.grid(True)
plt.annotate("gradien = v", xy=(2, x0+v*2), xytext=(3, x0+v*2+5),
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

**Hubungan antara Kecepatan dan Waktu $(v‚Äìt)$**

Pada Gerak Lurus Beraturan (GLB), kecepatan benda tetap selama bergerak. Tidak ada percepatan yang bekerja, sehingga grafik kecepatan‚Äìwaktu berbentuk garis datar atau horizontal.

**Persamaan hubungan $v‚Äìt$:**

$v(t)=v0$

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_velocity_time(v0=5.0, t_max=8.0, dt=0.1, show_examples=True, save_path=None):
    t = np.arange(0, t_max + dt/2, dt)
    v = np.full_like(t, fill_value=v0, dtype=float)

    plt.figure(figsize=(8,4))
    plt.plot(t, v, linewidth=2)
    plt.hlines(0, 0, t_max, colors='k', linewidth=0.8)
    plt.fill_between(t, v, 0, alpha=0.2)

    if show_examples:
        example_ts = np.array([0, 1, 2, t_max])
        plt.scatter(example_ts, np.full_like(example_ts, v0), zorder=5)
        for ti in example_ts:
            plt.annotate(f"t={ti:.0f}s\nv={v0:.1f} m/s",
                         xy=(ti, v0),
                         xytext=(ti, v0 + 0.5),
                         ha='center',
                         fontsize=9,
                         arrowprops=dict(arrowstyle='->', lw=0.7))

    plt.title("Kecepatan vs Waktu (GLB)")
    plt.xlabel("Waktu (s)")
    plt.ylabel("Kecepatan v (m/s)")
    plt.ylim(min(-0.1, v0 - 1), v0 + 2)
    plt.xlim(0, t_max)
    plt.grid(True, linestyle='--', linewidth=0.5)

    displacement = v0 * t_max
    plt.text(0.98 * t_max, v0 - 0.2, f"Perpindahan total = {displacement:.2f} m",
             ha='right', va='top', fontsize=10, bbox=dict(boxstyle="round,pad=0.3", alpha=0.15))

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

    plt.show()

if __name__ == "__main__":
    plot_velocity_time(v0=5.0, t_max=8.0, dt=0.05, show_examples=True, save_path=None)
```

**Hubungan antara Percepatan dan Waktu $(a‚Äìt)$**

Pada Gerak Lurus Beraturan (GLB), percepatan benda bernilai nol sepanjang waktu. Hal ini karena kecepatan tidak berubah (konstan). Akibatnya, grafik percepatan‚Äìwaktu berbentuk garis datar pada $a = 0.$

**Persamaan hubungan $a‚Äìt$:**

$v(t)=v0‚Äã$

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_acceleration_time(a_value=0.0, t_max=8.0, dt=0.1, show_examples=True, save_path=None):
    t = np.arange(0, t_max + dt/2, dt)
    a = np.full_like(t, fill_value=a_value, dtype=float)

    plt.figure(figsize=(8,4))
    plt.plot(t, a, linewidth=2)
    plt.hlines(0, 0, t_max, linewidth=0.8)
    plt.axhline(a_value, linestyle='--', linewidth=0.8)

    if show_examples:
        example_ts = np.array([0, 1, 2, t_max])
        plt.scatter(example_ts, np.full_like(example_ts, a_value), zorder=5)
        for ti in example_ts:
            plt.annotate(f"t={ti:.0f}s\n a={a_value:.1f} m/s¬≤",
                         xy=(ti, a_value),
                         xytext=(ti, a_value + 0.6 if a_value >= 0 else a_value - 0.6),
                         ha='center',
                         fontsize=9,
                         arrowprops=dict(arrowstyle='->', lw=0.6))

    plt.title("Percepatan vs Waktu (GLB)")
    plt.xlabel("Waktu (s)")
    plt.ylabel("Percepatan a (m/s¬≤)")
    ymin = min(-1.0, a_value - 1.0)
    ymax = max(1.0, a_value + 1.0)
    plt.ylim(ymin, ymax)
    plt.xlim(0, t_max)
    plt.grid(True, linestyle='--', linewidth=0.5)

    plt.text(0.02 * t_max, 0.85 * ymax,
             "GLB ‚Üí percepatan = 0\n(a(t) = 0 ‚áí kecepatan konstan)",
             fontsize=10, bbox=dict(boxstyle="round,pad=0.3", alpha=0.12))

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

    plt.show()

if __name__ == "__main__":
    plot_acceleration_time(a_value=0.0, t_max=8.0, dt=0.05, show_examples=True, save_path=None)
```

**Keterkaitan Grafik GLB (x‚Äìt, v‚Äìt, a‚Äìt)**

Pada Gerak Lurus Beraturan (GLB), percepatan benda bernilai nol sepanjang waktu. Hal ini karena kecepatan tidak berubah (konstan). Akibatnya, grafik percepatan‚Äìwaktu berbentuk garis datar pada $a = 0.$

**1. Grafik Posisi‚ÄìWaktu (x‚Äìt) ‚Üí Garis Miring**

- Menunjukkan perubahan posisi secara linear terhadap waktu.

- Kemiringan (gradien) grafik x‚Äìt sama dengan kecepatan v.

- Jika grafik semakin curam ‚Üí kecepatan semakin besar.

$ùë•(ùë°)=ùë•0+ùë£ùë°$

**2. Grafik Kecepatan‚ÄìWaktu (v‚Äìt) ‚Üí Garis Datar**

- Kecepatan tetap setiap saat.

- Grafik berupa garis horizontal pada nilai $ùë£0$.

- Luas di bawah kurva v‚Äìt menyatakan perpindahan.

    $v(t)=v0$
	‚Äã

**3. Grafik Percepatan‚ÄìWaktu (a‚Äìt) ‚Üí Garis Datar pada Nol**

- Percepatan bernilai nol karena tidak ada perubahan kecepatan.

- Grafik a‚Äìt selalu berada pada garis $a=0$.

  **a(t)=0**
  

**Keterkaitan Ketiganya**

| Grafik  | Memberi Informasi Tentang | Hubungannya                             |
| ------- | ------------------------- | --------------------------------------- |
| **x‚Äìt** | Perubahan posisi          | Kemiringannya = kecepatan (v)           |
| **v‚Äìt** | Kecepatan tetap           | Luas di bawah grafik = perubahan posisi |
| **a‚Äìt** | Percepatan nol            | Menjelaskan kenapa grafik v‚Äìt datar     |


```python
import numpy as np
import matplotlib.pyplot as plt

def velocity_update(v, dt):
    dv = 0.0
    return v, dv

def position_update(x, v, dt):
    dx = v * dt
    x_new = x + dx
    return x_new, dx

def acceleration_update():
    return 0.0

def glb_simulation(v0, x0=0.0, dt=0.1, t_max=10.0):
    x = x0
    v = v0
    a = acceleration_update()
    t = 0.0

    xs = [x]
    vs = [v]
    accs = [a]
    ts = [t]

    while t + dt <= t_max + 1e-12:
        v, dv = velocity_update(v, dt)
        x, dx = position_update(x, v, dt)
        a = acceleration_update()
        t += dt

        xs.append(x)
        vs.append(v)
        accs.append(a)
        ts.append(t)

    return xs, vs, accs, ts

def plot_position_time(ts, xs, example_points=None, save_path=None):
    plt.figure(figsize=(7,4))
    plt.plot(ts, xs, linewidth=2)
    if example_points is not None:
        for tt in example_points:
            plt.scatter(tt, xs[int(round(tt/(ts[1]-ts[0])))] , zorder=5)
            plt.annotate(f"t={tt:.1f}s\nx={xs[int(round(tt/(ts[1]-ts[0])))]:.2f} m",
                         xy=(tt, xs[int(round(tt/(ts[1]-ts[0])))]),
                         xytext=(tt, xs[int(round(tt/(ts[1]-ts[0])))] + 0.5),
                         ha='center',
                         fontsize=9,
                         arrowprops=dict(arrowstyle='->', lw=0.6))
    plt.title("Posisi vs Waktu (GLB)")
    plt.xlabel("Waktu (s)")
    plt.ylabel("Posisi x (m)")
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.xlim(min(ts), max(ts))
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

def plot_velocity_time(ts, vs, show_area=True, save_path=None):
    plt.figure(figsize=(7,3.5))
    plt.plot(ts, vs, linewidth=2)
    plt.hlines(0, ts[0], ts[-1], linewidth=0.8)
    if show_area:
        plt.fill_between(ts, vs, 0, alpha=0.18)
    plt.title("Kecepatan vs Waktu (GLB)")
    plt.xlabel("Waktu (s)")
    plt.ylabel("Kecepatan v (m/s)")
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.xlim(min(ts), max(ts))
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

def plot_acceleration_time(ts, accs, save_path=None):
    plt.figure(figsize=(7,3))
    plt.plot(ts, accs, linewidth=2)
    plt.hlines(0, ts[0], ts[-1], linewidth=0.8)
    plt.title("Percepatan vs Waktu (GLB)")
    plt.xlabel("Waktu (s)")
    plt.ylabel("Percepatan a (m/s¬≤)")
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.xlim(min(ts), max(ts))
    plt.ylim(min(-1.0, min(accs)-0.5), max(1.0, max(accs)+0.5))
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')

if __name__ == "__main__":
    v0 = 5.0
    x0 = 2.0
    dt = 0.1
    t_max = 8.0

    xs, vs, accs, ts = glb_simulation(v0, x0=x0, dt=dt, t_max=t_max)

    plot_position_time(ts, xs, example_points=[0,1,2], save_path="posisi_waktu.png")
    plot_velocity_time(ts, vs, show_area=True, save_path="kecepatan_waktu.png")
    plot_acceleration_time(ts, accs, save_path="percepatan_waktu.png")

    plt.show()

    print(f"Parameter: v0={v0} m/s, x0={x0} m, dt={dt} s, t_max={t_max} s")
    print(f"Posisi akhir: x({t_max}s) = {xs[-1]:.3f} m")
    print(f"Perpindahan total (analitik) = v0 * t_max = {v0 * t_max:.3f} m")
```
