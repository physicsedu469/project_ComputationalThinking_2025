# **Simulasi Lensa Cembung â€“ Bayangan Benda**

Lensa cembung merupakan jenis lensa yang lebih tebal di bagian tengah dibandingkan tepinya.
Sifat utama lensa ini adalah memfokuskan berkas cahaya sejajar ke satu titik fokus.
Apabila sebuah benda ditempatkan di depan lensa, maka lensa akan membentuk bayangan benda tersebut
dengan karakteristik tertentu, bergantung pada posisi benda terhadap lensa.

## Rumus Lensa Tipis

Hubungan antara jarak benda ($s$), jarak bayangan ($s'$), dan panjang fokus lensa ($f$) dinyatakan oleh persamaan lensa tipis:

$$
\frac{1}{f} = \frac{1}{s} + \frac{1}{s'}
$$

Persamaan ini diperoleh dari analisis geometri sinar pada lensa tipis dengan pendekatan paraxial
(sinar membentuk sudut kecil terhadap sumbu utama). Dengan menelusuri sinar sejajar, sinar yang
melalui pusat optik, serta sinar yang melalui fokus, terbentuk segitiga-segitiga sebangun yang
menghasilkan hubungan matematis di atas.

## Interpretasi Tanda

- Jika $s' > 0$, bayangan bersifat **nyata** dan terbentuk di sisi berlawanan dari benda.
  Bayangan nyata dapat ditangkap pada layar dan biasanya terbalik.
- Jika $s' < 0$, bayangan bersifat **maya** dan terbentuk di sisi yang sama dengan benda.
  Bayangan maya tidak dapat ditangkap layar, tetapi dapat dilihat melalui lensa dan tampak tegak.

## Pembesaran Bayangan

Selain posisi bayangan, sifat bayangan juga ditentukan oleh pembesaran ($M$):

$$
M = -\frac{s'}{s}
$$

- Jika $M < 0$, bayangan nyata dan terbalik.  
- Jika $M > 0$, bayangan maya dan tegak.  

Simulasi lensa cembung tidak hanya menunjukkan posisi bayangan,
tetapi juga bagaimana ukuran dan orientasi bayangan berubah sesuai dengan jarak benda terhadap lensa.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter lensa
f = 10.0  # panjang fokus (cm)

# Rentang s di kedua sisi fokus, hindari s = f (singular)
s_left  = np.linspace(5.0, 9.9, 400)    # s < f (bayangan maya)
s_right = np.linspace(10.1, 100.0, 600) # s > f (bayangan nyata)

def s_prime_from(s, f):
    # 1/f = 1/s + 1/s'  ->  s' = 1 / (1/f - 1/s)
    return 1.0 / (1.0/f - 1.0/s)

# Hitung s' dan M untuk kedua sisi
sprime_left  = s_prime_from(s_left, f)
sprime_right = s_prime_from(s_right, f)

M_left  = -sprime_left / s_left
M_right = -sprime_right / s_right
```

```python
# --- Grafik 1: Posisi bayangan vs posisi benda ---
plt.figure(figsize=(9,6))
plt.plot(s_left,  sprime_left,  color='royalblue', label="s' (s < f, bayangan maya)")
plt.plot(s_right, sprime_right, color='navy',      label="s' (s > f, bayangan nyata)")
plt.axvline(f, color='gray', linestyle='--', label=f'Fokus (f = {f} cm)')
plt.title("Posisi Bayangan vs Posisi Benda (Lensa Cembung)")
plt.xlabel("Jarak benda s (cm)")
plt.ylabel("Jarak bayangan s' (cm)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Grafik 2: Pembesaran vs posisi benda ---
plt.figure(figsize=(9,6))
plt.plot(s_left,  M_left,  color='darkgreen', label="M (s < f, maya, tegak)")
plt.plot(s_right, M_right, color='green',     label="M (s > f, nyata, terbalik)")
plt.axvline(f, color='gray', linestyle='--', label=f'Fokus (f = {f} cm)')
plt.axhline(0, color='black', linewidth=0.8)
plt.title("Pembesaran Bayangan vs Posisi Benda")
plt.xlabel("Jarak benda s (cm)")
plt.ylabel("Pembesaran M")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
