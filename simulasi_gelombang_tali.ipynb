**Pendahuluan**

Gelombang pada tali adalah gelombang mekanik yang merambat sepanjang tali. Pada simulasi ini kita akan memvisualisasikan:


*   Bentuk gelombang berjalan
*   Perubahan posisi simpul dan perut

*   Pengaruh amplitudo (A), frekuensi (f), dan cepat rambat (v)
*   Hubungan antara fungsi posisi dan waktu

Gelombang yang akan dianalisis adalah gelombang berjalan satu dimensi dengan bentuk umum:


**Rumus dasar gelombang tali**

Gelombang berjalan dapat dideskripsikan dengan:

y (x, t) = A sin⁡ (kx-ωt)

dengan:

A : amplitudo (meter)

k : 2π/⋋ = bilangan gelombang

ω : 2πf = frekuatan sudut

v = f⋋ = cepat rambat gelombang







```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter dasar gelombang
A = 1         # amplitudo (meter)
lam = 2       # panjang gelombang (meter)
f = 1         # frekuensi (Hz)
v = f * lam   # cepat rambat
```

```python
k = 2*np.pi/lam   # bilangan gelombang
omega = 2*np.pi*f # frekuensi sudut
```

```python
# domain posisi dan waktu
x = np.linspace(0, 10, 500)
t = 0   # waktu awal
```

```python
y = A * np.sin(k*x - omega*t)
```

```python
plt.figure(figsize=(10,4))
plt.plot(x, y)
```

```python
plt.title("Gelombang Berjalan pada Tali (t = 0 s)")
plt.xlabel("Posisi x (meter)")
plt.ylabel("Perpindahan y (meter)")
plt.grid(True)
plt.show()
```

```python
t_values = np.linspace(0, 2, 200)
```

```python
plt.figure(figsize=(10,4))
for t in [0, 0.5, 1, 1.5, 2]:
    y = A * np.sin(k*x - omega*t)
    plt.plot(x, y, label=f"t = {t} s")
```

```python
plt.title("Perubahan Gelombang terhadap waktu")
plt.xlabel("x (meter)")
plt.ylabel("y (meter)")
plt.legend()
plt.grid(True)
plt.show
```

```python
print("Bilangan gelombang (k)   =", k, "rad/m")
print("Frekuensi sudut (omega)  =", omega, "rad/s")
print("Cepat rambat (v)         =", v, "m/s")
```

```python
import ipywidgets as widgets
from ipywidgets import interactive
```

```python
def gelombang(A=1.0, lam=2.0, f=1.0, t=0.0):
    v = f * lam
    k = 2*np.pi/lam
    omega = 2*np.pi*f
```

```python
x = np.linspace(0, 10, 500)
y = A * np.sin(k*x - omega*t)
```

```python
plt.figure(figsize=(10,4))
plt.plot(x, y)
plt.ylim(-A*1.5, A*1.5)
plt.title(f"Gelombang pada tali (A={A}, lam={lam}, f={f}, t={t})")
plt.xlabel("x (meter)")
plt.ylabel("y (meter)")
plt.grid(True)
plt.show()
```

```python
interactive(gelombang)
A=(0.5, 3.0, 0.1),
lam=(0.5, 5.0, 0.1),
f=(0.5, 5.0, 0.1),
t=(0, 5, 0.5)
```

**Kesimpulan**

Dari simulasi di atas dapat ditarik kesimpulan sebagai berikut:


1.   Gelombang merambat ke arah +x dengan pola sinus.
2.   Semakin besar frekuensi, semakin cepat perubahan bentuk gelombang terhadap waktu.
3.   Semakin besar panjang gelombang, semakin jarang pola puncak-lembah.
4.   Amplitudo menentukan tinggi maksimum gelombang.
